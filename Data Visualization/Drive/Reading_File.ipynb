```python
#groupby concept
#1. Average Price by Manufacturer
avg_price = df.groupby('Manufacturer')['Price'].sum()
print(avg_price)
```

```python
#2. Total Cars Sold (Count) by Manufacturer
car_count = df.groupby('Manufacturer')['Model'].count()
print(car_count)
```

```python
#3. Average Mileage by Fuel Type
avg_mileage = df.groupby('Fuel type')['Mileage'].mean().reset_index()
print(avg_mileage)
```

```python
#4.Multiple Aggregations (Mean and Max Price)
price_stats = df.groupby('Manufacturer')['Price'].agg(['mean', 'max','min'])
print(price_stats)
```

```python
#5. Group by Two Columns (Manufacturer & Fuel Type)
group_two = df.groupby(['Manufacturer', 'Fuel type'])['Price'].mean()
print(group_two)
```

```python
#6. Sorting the Grouped Results
avg_price_sorted = df.groupby('Manufacturer')['Price'].mean().sort_values(ascending=False)
print(avg_price_sorted.head(10))
```

```python
#7. Count and Average Together
summary = df.groupby('Manufacturer').agg(
    total_cars=('Model', 'count'),
    avg_price=('Price', 'mean')
)
print(summary)
```

```python
#8. Displaying Results as a DataFrame
summary_df = df.groupby('Fuel type')[['Mileage', 'Price']].mean().reset_index()
print(summary_df)
```

```python
top_index = df.groupby('Manufacturer')['Price'].idxmin()
top_cars = df.loc[top_index][['Manufacturer', 'Model', 'Price','Fuel type']]
print(top_cars)
print("====")
# print(top_index)
```

```python
import pandas as pd

data = {
    'Date': ['2025-01-01', '2025-01-01', '2025-01-02', '2025-01-02', '2025-01-03', '2025-01-03', '2025-01-04'],
    'Region': ['East', 'East', 'West', 'South', 'East', 'West', 'South'],
    'Product': ['Pen', 'Notebook', 'Pencil', 'Mug', 'Mug', 'Pen', 'Notebook'],
    'Category': ['Stationery', 'Stationery', 'Stationery', 'Gift', 'Gift', 'Stationery', 'Stationery'],
    'Sales': [120, 200, 100, 250, 300, 90, 180],
    'Quantity': [10, 5, 8, 4, 6, 7, 3]
}

df = pd.DataFrame(data)
print(df)
```

```python
#Total Sales by Region
pivot1 = pd.pivot_table(df, values='Sales', index='Region', aggfunc='sum')
print(pivot1)
```

```python
#Sales by Region and Category
pivot2 = pd.pivot_table(df, values='Sales', index='Region', columns='Category', aggfunc='sum', fill_value=0)
print(pivot2)
```

```python
#Multiple Aggregations (Sum and Mean)
pivot3 = pd.pivot_table(
    df,
    values='Sales',
    index='Region',
    columns='Category',
    aggfunc=['sum', 'mean'],
    fill_value=0
)
print(pivot3)
```

```python
#Analyze Total Quantity per Product
pivot4 = pd.pivot_table(df, values='Quantity', index='Product', aggfunc='sum')
print(pivot4)
```

```python
#Combine Multiple Values
pivot5 = pd.pivot_table(
    df,
    values=['Sales', 'Quantity'],
    index='Region',
    columns='Category',
    aggfunc='sum',
    fill_value=0
)
print(pivot5)
```

```python
pivot5.columns = ['_'.join(col).strip() for col in pivot5.columns.values]
print(pivot5)
```
