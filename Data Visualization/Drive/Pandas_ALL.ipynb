```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Name": ["Amit", "Bhavna", "Chetan", "Disha", "Esha"],
    "Age": [25, 30, 35, 40, 28],
    "City": ["Mumbai", "Delhi", "Mumbai", "Pune", "Pune"],
    "Salary": [50000, 60000, 55000, 65000, 70000],
    "Department": ["Sales", "HR", "Sales", "IT", "HR"]
}
df = pd.DataFrame(data)
df
```

```python
# ============================================================
# 4. SELECTION & INDEXING: loc, iloc, column selection
# ============================================================
# Column selection:
#   df["col"] -> Series
#   df[["col1","col2"]] -> DataFrame
```

```python
# 4.1 Selecting columns
print("Single column (Series):\n", df["Name"])
print("\nMultiple columns (DataFrame):\n", df[["Name", "Salary"]])
```

```python
# 4.2 .loc -> label-based selection (rows, columns)
# Syntax: df.loc[row_labels, column_labels]

# Single row by label (here default index 0..n-1)
print("\nRow with index label 0 using loc:\n", df.loc[0])
```

```python
# Multiple rows and selected columns
subset_loc = df.loc[0:2, ["Name", "City", "Salary"]]  # rows 0 to 2 inclusive
print("\nSubset using loc:\n", subset_loc)
```

```python
# 4.3 .iloc -> integer position-based selection
# Syntax: df.iloc[row_positions, column_positions]
print("\nRow at position 1 using iloc:\n", df.iloc[1])
```

```python
subset_iloc = df.iloc[1:4, 1:4]  # rows 1-3, cols 1-3
print("\nSubset using iloc:\n", subset_iloc)
```

```python
# 4.4 Boolean indexing
# Example: Employees with Salary > 60000
high_salary = df[(df["Age"] != 20) & (df["Salary"]!=70000)]
print("\nEmployees with high salary:\n", high_salary)
```

```python
# Combine multiple conditions using & (and) / | (or)
# Example: Salary > 55000 AND City == 'Pune'
cond = (df["Salary"] > 55000) & (df["City"] == "Pune")
print("\nSalary > 55000 and City Pune:\n", df[cond])
```

```python
# ============================================================
# 5. FILTER, SORT, RENAME, DROP, ASSIGN
# ============================================================
```

```python
# 5.1 .sort_values()
# Parameters:
#   by : column name or list of columns
#   ascending : True/False or list
#   inplace : modify original DataFrame if True
```

```python
# 5.3 .drop()
# Parameters:
#   labels : name(s) to drop
#   axis : 0 for rows, 1 for columns
#   inplace : True/False
```

```python
# Drop a column
df_drop_col = df.drop(columns=["Department"])
print("\nDropped 'Department' column:\n", df_drop_col.head())
```

```python
# 5.4 .assign() -> create new columns without modifying original
# Parameters: **kwargs as new_col = expression
df_new = df.assign(
    Salary_in_Lakhs = df["Salary"] / 100000,
    Age_Category = np.where(df["Age"] >= 30, "Senior", "Junior")
)
print("\nNew columns with assign:\n", df_new)
```

```python
# ============================================================
# 6. HANDLING MISSING DATA: isna, notna, fillna, dropna
# ============================================================
```

```python
# Let's create a small DataFrame with NaNs
df_na = pd.DataFrame({
    "A": [1, 2, np.nan, 4],
    "B": [np.nan, 5, 6, 7],
    "C": ["x", "y", "z", None]
})

print("Original with NaNs:\n", df_na)
```

```python
# .isna() / .isnull() -> True where value is NaN
print("\nIs NaN:\n", df_na.isna().sum())

# .notna() / .notnull() -> True where value is NOT NaN
print("\nNot NaN:\n", df_na.notna())
```

```python
# .fillna()
# Parameters:
#   value : scalar, dict, Series, or DataFrame of replacement values
#   method : 'ffill' (forward fill), 'bfill' (backward fill)
```

```python
# 6.1 Fill with a scalar value
print("\nFill NaNs with 0:\n", df_na.fillna(0))

# 6.2 Fill using column-wise dictionary
fill_values = {"A": df_na["A"].mean(), "B": 0, "C": "missing"}
print("\nFill NaNs with column-wise dict:\n", df_na.fillna(fill_values))

# 6.3 Forward fill (propagate last valid value downwards)
print("\nForward fill:\n", df_na.ffill())
```

```python
# .dropna()
# Parameters:
#   axis : 0 (rows) or 1 (columns)
#   how : 'any' (default) or 'all'
#   subset : list of columns to consider
```

```python
# 6.4 Drop rows where ANY NaN
print("\nDrop rows with any NaN:\n", df_na.dropna())

# 6.5 Drop columns where ALL values are NaN
df_all_na = pd.DataFrame({"A": [1, np.nan], "B": [np.nan, np.nan]})
print("\nDrop columns where all NaN:\n", df_all_na.dropna(axis=1, how="all"))
```

```python
# 9.1 pd.merge() -> SQL-style joins
# Parameters:
#   left, right : DataFrames
#   on : column(s) to join on
#   how : 'inner', 'left', 'right', 'outer'
```

```python
employees = pd.DataFrame({
    "EmpID": [1, 2, 3, 4],
    "Name": ["Amit", "Bhavna", "Chetan", "Disha"],
    "DeptID": [10, 20, 10, 30]
})

departments = pd.DataFrame({
    "DeptID": [10, 20, 40],
    "Department": ["Sales", "HR", "Finance"]
})

print("Employees:\n", employees)
print("\nDepartments:\n", departments)
```

```python
# Inner join (common DeptID only)
inner_join = pd.merge(employees, departments, on="DeptID", how="inner")
print("\nInner join:\n", inner_join)
```

```python
# Left join (all employees, dept info if available)
left_join = pd.merge(employees, departments, on="DeptID", how="left")
print("\nLeft join:\n", left_join)
```

```python
# Right join (all departments, employee info if available)
right_join = pd.merge(employees, departments, on="DeptID", how="right")
print("\nRight join:\n", right_join)
```

```python
# Outer join (all DeptID from both sides)
outer_join = pd.merge(employees, departments, on="DeptID", how="outer")
print("\nOuter join:\n", outer_join)
```

```python
# 9.2 pd.concat() -> concatenate DataFrames/Series vertically or horizontally
# Parameters:
#   objs : list of DataFrames/Series
#   axis : 0 (rows) or 1 (columns)
#   ignore_index : reindex if True
```

```python
df_top = df.iloc[0:2]
df_bottom = df.iloc[2:5]
print("1st",df_top)
print("----")
print("2nd",df_bottom)
```

```python
vertical_concat = pd.concat([df_top, df_bottom], axis=0)
print("\nVertical concat (rows):\n", vertical_concat)
```

```python
# Horizontal concat (join columns side by side)
horizontal_concat = pd.concat(
    [df_top.reset_index(drop=True), df_bottom.reset_index(drop=True)],
    axis=1
)
print("\nHorizontal concat (columns):\n", horizontal_concat)
```

```python
# Concatenate columns side by side
extra = pd.DataFrame({"Bonus": [5000, 6000, 7000, 8000, 9000]})
horizontal_concat = pd.concat([df.reset_index(drop=True), extra], axis=1)
print("\nHorizontal concat (columns):\n", horizontal_concat)
```
