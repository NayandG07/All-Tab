```python
import numpy as np
import time
```

```python
# Example 1: Elementwise add (lists vs NumPy)
lst1 = list(range(1_00_000))
lst2 = list(range(1_00_000, 2_00_000))

start = time.time()
lst_add = [x + y for x, y in zip(lst1, lst2)]
t_list = time.time() - start

arr1 = np.arange(1_00_000)
arr2 = np.arange(1_00_000, 2_00_000)

start = time.time()
arr_add = arr1 + arr2
t_np = time.time() - start

print(f"List time: {t_list:.6f}s, NumPy time: {t_np:.6f}s")
```

```python
# Example 2: Scalar multiply
start = time.time()
lst_scaled = [x * 3 for x in lst1]
t_list2 = time.time() - start

start = time.time()
arr_scaled = arr1 * 3
t_np2 = time.time() - start

print(f"List x3 time: {t_list2:.6f}s, NumPy x3 time: {t_np2:.6f}s")
```

```python
# Example 3: Sum of squares
start = time.time()
sum_sq_list = sum([x*x for x in lst1])
t_list3 = time.time() - start

start = time.time()
sum_sq_np = np.sum(arr1 * arr1)
t_np3 = time.time() - start

print(f"List sum of squares: {t_list3:.6f}s, NumPy: {t_np3:.6f}s")
```

```python
# Example 4: Memory footprint
l_small = [1,2,3,4,5]
a_small = np.array(l_small, dtype=np.int64)
import sys
print("Approx list bytes:", sum(sys.getsizeof(x) for x in l_small) + sys.getsizeof(l_small))
print("NumPy nbytes:", a_small.nbytes)
```

**Note: Times vary by machine, but NumPy will usually be much faster.**

```python
# ---- From Python sequences ----
# Example 1: From list (1D)
a1 = np.array([1, 2, 3, 4])
print("From list:", a1, "| dtype:", a1.dtype)
```

```python
# Example 2: From nested list (2D matrix)
a2 = np.array([[1, 2, 3], [4, 5, 6]])
print("2D array:\n", a2)
```

```python
# Example 3: From tuple with dtype
a3 = np.array((1.1, 2.2, 3.3), dtype=np.float32)
print("From tuple with dtype:", a3, "| dtype:", a3.dtype)
```

```python
# ---- Range-like builders ----
# Example 4: arange(start, stop, step)
r1 = np.arange(0, 10, 2)  # 0..8 step 2
print("arange:", r1)
```

```python
# Example 5: linspace(start, stop, num) inclusive endpoints
r2 = np.linspace(0,71)
print("linspace:", r2)
print(len(r2))
```

```python
# Example 6: logspace(start_exp, stop_exp, num) -> 10**exponent
r3 = np.logspace(0, 3, 3,base=9)  # 10^0 .. 10^3 with 4 points
print("logspace:", r3)
```

```python
# ---- Special arrays ----
# Example 7: zeros / ones / full
import numpy as np
z = np.zeros((3,4))
o = np.ones((2, 3))
f = np.full((2, 3), fill_value="bhi")
print("zeros:\n", z)
print("ones:\n", o)
print("full(7):\n", f)
```

```python
# Example 8: Identity/eye + diagonal
I = np.eye(5)         # Identity
D = np.diag([5, 3, 2,12,54])  # Diagonal from 1D list
print("Identity:\n", I)
print("Diagonal:\n", D)
```

```python
# Example 9: Random-based creation
np.random.seed(41)
rand_uniform = np.random.rand(2, 3)         # [0,1)
rand_normal  = np.random.randn(2, 3)        # standard normal
rand_ints    = np.random.randint(10, 50, size=(2,3))
print("rand uniform:\n", rand_uniform)
print("rand normal:\n", rand_normal)
print("rand ints:\n", rand_ints)
```

```python
# ---- 1D ----
x = np.array([10, 20, 30, 40, 50, 60])
x
```

```python
# Example 1: Basic indexing
print("x[0] =", x[0], "| x[-1] =", x[-1])
```

```python
# Example 2: Slicing ranges
print("x[1:4] =", x[1:4])
print("x[:3]  =", x[:3])
print("x[3:]  =", x[3:])
```

```python
# Example 3: Step & reverse
print("x[::2]  =", x[::2])    # every 2nd
print("x[::-1] =", x[::-1])   # reverse
```

```python
# ---- 2D ----
A = np.arange(1, 13).reshape(3, 4)
A
```

```python
# Example 4: Index specific item
print("A[0, 2] =", A[0,2])      # row 0, col 2 -> 3
```

```python
# Example 5: Entire row/col
print("Row 1:", A[1, :])         # 2nd row
print("Col 2:", A[:, 2])         # 3rd column
```

```python
# Example 6: Submatrix slice
print("Block A[0:2, 1:3]:\n", A[0:2, 1:3])  # rows 0-1, cols 1-2
```

```python
# ---- 3D ----
B = np.arange(2*3*4).reshape(2, 3, 4)  # (depth=2, rows=3, cols=4)
B
```

```python
# Example 7: Indexing across 3 axes
print("B[1, 2, 3] =", B[1, 2, 3])
```

```python
# Example 8: Slice along last axis
print("B[:, 1, :]:\n", B[:, 1, :])
```

```python
# Example 9: Slice a 3D block
print("B[0:2, 0:2, 2:4]:\n", B[0:2, 0:2, 2:4])
```
