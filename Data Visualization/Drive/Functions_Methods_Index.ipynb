- <span style="color:#60a5fa;"><b>np.arange</b></span>: <span style="color:#facc15;"><i>Create evenly spaced values within a given interval (often used for index-like sequences).</i></span>
- <span style="color:#60a5fa;"><b>np.argsort</b></span>: <span style="color:#facc15;"><i>Return indices that would sort an array.</i></span>
- <span style="color:#60a5fa;"><b>np.array</b></span>: <span style="color:#facc15;"><i>Convert input data (like list/tuple) into a NumPy array.</i></span>
- <span style="color:#60a5fa;"><b>np.array_split</b></span>: <span style="color:#facc15;"><i>Split an array into multiple sub-arrays, allowing unequal sizes.</i></span>
- <span style="color:#60a5fa;"><b>np.concatenate</b></span>: <span style="color:#facc15;"><i>Join arrays along an existing axis.</i></span>
- <span style="color:#60a5fa;"><b>np.diag</b></span>: <span style="color:#facc15;"><i>Create a diagonal array or extract the diagonal from a matrix.</i></span>
- <span style="color:#60a5fa;"><b>np.dstack</b></span>: <span style="color:#facc15;"><i>Stack arrays depth-wise (along the third axis).</i></span>
- <span style="color:#60a5fa;"><b>np.exp</b></span>: <span style="color:#facc15;"><i>Compute element-wise exponential (e^x).</i></span>
- <span style="color:#60a5fa;"><b>np.eye</b></span>: <span style="color:#facc15;"><i>Create a 2D identity matrix.</i></span>
- <span style="color:#60a5fa;"><b>np.flatten</b></span>: <span style="color:#facc15;"><i>Return a flattened copy of an array as 1D.</i></span>
- <span style="color:#60a5fa;"><b>np.full</b></span>: <span style="color:#facc15;"><i>Create an array of a given shape filled with a constant value.</i></span>
- <span style="color:#60a5fa;"><b>np.hsplit</b></span>: <span style="color:#facc15;"><i>Split an array horizontally (column-wise).</i></span>
- <span style="color:#60a5fa;"><b>np.hstack</b></span>: <span style="color:#facc15;"><i>Stack arrays horizontally (column-wise for 2D arrays).</i></span>
- <span style="color:#60a5fa;"><b>np.insert</b></span>: <span style="color:#facc15;"><i>Insert values into an array at specified positions.</i></span>
- <span style="color:#60a5fa;"><b>np.linspace</b></span>: <span style="color:#facc15;"><i>Generate evenly spaced numbers over a specified range.</i></span>
- <span style="color:#60a5fa;"><b>np.logspace</b></span>: <span style="color:#facc15;"><i>Generate numbers spaced evenly on a log scale.</i></span>
- <span style="color:#60a5fa;"><b>np.maximum</b></span>: <span style="color:#facc15;"><i>Element-wise maximum of two arrays.</i></span>
- <span style="color:#60a5fa;"><b>np.mean</b></span>: <span style="color:#facc15;"><i>Compute arithmetic mean along an axis or for all values.</i></span>
- <span style="color:#60a5fa;"><b>np.minimum</b></span>: <span style="color:#facc15;"><i>Element-wise minimum of two arrays.</i></span>
- <span style="color:#60a5fa;"><b>np.nonzero</b></span>: <span style="color:#facc15;"><i>Return indices of non-zero elements.</i></span>
- <span style="color:#60a5fa;"><b>np.ones</b></span>: <span style="color:#facc15;"><i>Create an array filled with ones.</i></span>
- <span style="color:#60a5fa;"><b>np.ravel</b></span>: <span style="color:#facc15;"><i>Return a flattened 1D view when possible (copy only if needed).</i></span>
- <span style="color:#60a5fa;"><b>np.random.rand</b></span>: <span style="color:#facc15;"><i>Generate random samples from a uniform distribution in [0, 1).</i></span>
- <span style="color:#60a5fa;"><b>np.random.randint</b></span>: <span style="color:#facc15;"><i>Generate random integers in a specified range.</i></span>
- <span style="color:#60a5fa;"><b>np.random.randn</b></span>: <span style="color:#facc15;"><i>Generate random samples from a standard normal distribution.</i></span>
- <span style="color:#60a5fa;"><b>np.random.seed</b></span>: <span style="color:#facc15;"><i>Set random seed for reproducible random results.</i></span>

```python
book = pd.DataFrame({
    "Book_ID": [1, 2, 3, 4, 5],
    "Title": ["Python Basics", "Data Science 101", "AI Fundamentals", "Machine Learning", "SQL Guide"],
    "Category": ["Programming", "Data Science", "AI", "AI", "Database"],
    "Pages": [320, 410, 380, 450, 300],
    "Rating": [4.5, 4.7, 4.6, 4.8, 4.2]
})

book
```

```python
book[book["Rating"] > 4.5]
```

```python
book[book["Category"] == "AI"]
```

```python
book.groupby("Category")["Rating"].mean()
```

```python
book.sort_values("Pages", ascending=False).head(1)
```

```python
book.value_counts("Category")
```

```python
book.sort_values("Rating", ascending=False)
```

```python
book = book.assign(
    Difficulty_Level = np.where(book["Pages"] > 400, "Advanced", "Beginner")
)
book
```

```python
book = book.assign(
    Difficulty_Level=book["Pages"].apply(lambda x: "Advanced" if x > 400 else "Beginner")
)
book
```
