```python
# Example 1: Elementwise between equal-shaped arrays
a = np.array([2, 4, 6, 8], dtype=np.int32)
b = np.array([1, 3, 5, 7], dtype=np.int32)
print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)    # float division
```

```python
# Example 2: Scalar with array (applies to each element)
print("a * 10 =", a * 10)
print("b ** 2 =", b ** 2)
```

```python
# Example 3: Integer division, modulus, power
c = np.array([7, 10, 13])
d = np.array([3, 5, 2])
print("c // d =", c // d)  # floor division
print("c % d  =", c % d)   # remainder
print("c ** d =", c ** d)  # power
```

```python
# Example 4: Comparisons produce boolean arrays (useful for masking)
x = np.array([5, 12, 9, 21])
print("x > 10:", x > 10)
print("x == 9:", x == 9)
print("x <= 5:", x <= 5)
```

```python
# Example 2: Aggregation ufuncs (on axis)
A = np.arange(1, 13).reshape(3, 4)
print("A:\n", A)
print("sum all:", np.sum(A))
print("sum by rows (axis=1):", np.sum(A, axis=1))
print("max by cols (axis=0):", np.mean(A, axis=0))
```

```python
# Example 1: Reshape 1D -> 2D and back
x = np.arange(12)         # [0..11]
X = x.reshape(3, 4)       # 3 rows, 4 cols
print("X:\n", X)
print("Back to 1D (ravel):", X.ravel())
```

```python
import numpy as np
# Example 2: Using -1 to infer dimension
Y = np.arange(24).reshape(-1, 2)  # 2 rows, infer columns
print("Y shape:", Y.shape)
Z = np.arange(24).reshape(3, -1, 4)  # infer number of blocks
print("Z shape:", Z.shape)
```

```python
# Example 3: flatten vs ravel (copy vs view)
A = np.arange(6).reshape(2, 3)
r = A.ravel()
f = A.flatten()
A[0, 0] = 999
print("After modifying A[0,0]=999")
print("A:\n", A)
print("ravel view (r):", r)     # may reflect change
print("flatten copy (f):", f)   # not affected
```

```python
# Example 4: Reshape for matrix operations (column/row vectors)
v = np.array([1, 2, 3, 4])
col = v.reshape(-1, 1)
row = v.reshape(1, -1)
print("col shape:", col.shape, "\n", col)
print("row shape:", row.shape, "\n", row)
```

```python
# Example 1: hstack/vstack/concatenate
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("hstack:", np.hstack([a, b]))
print("vstack:\n", np.vstack([a, b]))
print("concatenate axis=0:", np.concatenate([a, b], axis=0))
```

```python
# Example 2: 2D stacking
A = np.arange(1, 7).reshape(2, 3)
B = np.arange(7, 13).reshape(2, 3)
print("A:\n", A)
print("B:\n", B)
print("hstack 2D:\n", np.hstack([A, B]))  # concat columns
print("vstack 2D:\n", np.vstack([A, B]))  # concat rows
```

```python
# Example 4: Splitting (hsplit/vsplit/array_split)
M = np.arange(16).reshape(4, 4)
print("M:\n", M)
print("hsplit into 2:\n", np.hsplit(M, 4))
print("vsplit into 2:\n", np.vsplit(M, 4))
print("array_split uneven cols (3 parts):")
print(np.array_split(M, 3, axis=1))  # flexible (last block smaller)
```

```python
x = np.array(["a", "b", "f", "a"])
print("sorted copy:", np.sort(x))
print("argsort (indices):", np.argsort(x))
```

```python
# Example 1: sort vs argsort
x = np.array([40, 10, 30, 20])
print("sorted copy:", np.sort(x))
print("argsort (indices):", np.argsort(x))
print("reorder using argsort:", x[np.argsort(x)])
```

```python
# Example 2: sort 2D by rows/cols
A = np.array([[3, 1, 8],
              [9, 7, 2]])
print("sort rows (axis=1):\n", np.sort(A, axis=1))
print("sort cols (axis=0):\n", np.sort(A, axis=0))
```

```python
# Example 3: where / nonzero
y = np.array([5, 0, 7, 0, 9, 0])
print("where y==0 ->", np.where(y!=0))
print("nonzero ->", np.nonzero(y))
```

```python
# Example 4: searchsorted (binary search on sorted arrays)
sorted_arr = np.array([2, 4, 6, 8, 10])
print("insert positions for [1,5,9]:", np.searchsorted(sorted_arr, [1, 5, 9]))
print("left vs right side for equal values:")
dup = np.array([1, 2, 2, 2, 3])
print("left:", np.searchsorted(dup, 2, side='left'))
print("right:", np.searchsorted(dup, 2, side='right'))
```

```python
import numpy as np

sorted_arr = np.array([2, 4, 6, 8, 10])

# Get positions where new values should be inserted
values = [1, 5, 9]
positions = np.searchsorted(sorted_arr, values)

print("Positions:", positions)

# Insert values at those positions
new_arr = np.insert(sorted_arr, positions, values)
print("Array after insertion:", new_arr)
```
