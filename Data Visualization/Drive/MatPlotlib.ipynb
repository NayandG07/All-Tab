```python
# -------------------------------------------------------------
# 1. SETUP & LOAD DATASET
# -------------------------------------------------------------
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

# Load the iris dataset into a pandas DataFrame
iris = load_iris(as_frame=True)
df = iris.frame

# Display first 5 rows
print("Dataset preview:")
display(df.head())

print("\nColumns:", list(df.columns))
print("Target classes:", iris.target_names)
```

```python
# -------------------------------------------------------------
# 2. LINE PLOT
# -------------------------------------------------------------
# Line plots are used when data points are connected — showing change or trend.

plt.plot(df.index, df["sepal length (cm)"], color='blue')

plt.title("Line Plot - Sepal Length over Samples")
plt.xlabel("Sample Index")
plt.ylabel("Sepal Length (cm)")
plt.grid(True)
plt.show()

# 💡 Explanation:
# Each point represents one flower’s sepal length.
# The line connects them in order, showing variation or trend.
# Useful for time series or ordered data.
```

```python
# -------------------------------------------------------------
# 3. SCATTER PLOT
# -------------------------------------------------------------
# Scatter plots show relationship between two numerical features.

plt.scatter(df["sepal length (cm)"], df["sepal width (cm)"], color='green')

plt.title("Scatter Plot - Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.grid(True)
plt.show()

# 💡 Explanation:
# Each point is one flower.
# It helps us see how sepal length and width relate.
# Useful for identifying patterns, trends, or correlations.
```

```python
# -------------------------------------------------------------
# 4. BAR PLOT
# -------------------------------------------------------------
# Bar plots are used for comparing categories.
# Here we calculate the average petal length for each flower species.

species_avg = df.groupby("target")["petal length (cm)"].mean()
species_names = iris.target_names

plt.bar(species_names, species_avg, color=['red', 'orange', 'green'])

plt.title("Bar Plot - Average Petal Length per Species")
plt.xlabel("Species")
plt.ylabel("Average Petal Length (cm)")
plt.show()

# 💡 Explanation:
# Each bar represents the average petal length for that species.
# Bar charts are best for categorical comparisons.
```

```python
# -------------------------------------------------------------
# 5. HORIZONTAL BAR PLOT
# -------------------------------------------------------------
# Same as bar chart, but flipped horizontally.

plt.barh(species_names, species_avg, color=['skyblue', 'pink', 'lime'])
plt.title("Horizontal Bar Plot - Petal Length per Species")
plt.xlabel("Average Petal Length (cm)")
plt.ylabel("Species")
plt.show()

# 💡 Explanation:
# Useful when category names are long or you want a different visual style.
```

```python
# -------------------------------------------------------------
# 6. HISTOGRAM
# -------------------------------------------------------------
# Histograms show how frequently data values occur.

plt.hist(df["petal width (cm)"], bins=15, color='purple')

plt.title("Histogram - Petal Width Distribution")
plt.xlabel("Petal Width (cm)")
plt.ylabel("Frequency")
plt.show()

# 💡 Explanation:
# Shows how many flowers have certain petal widths.
# Useful for understanding data distribution (normal, skewed, etc.).
```

```python
# -------------------------------------------------------------
# 7. BOX PLOT
# -------------------------------------------------------------
# Box plots show the spread, median, and outliers of data.

plt.boxplot(df["petal length (cm)"], patch_artist=True, boxprops=dict(facecolor="lightgreen"))
plt.title("Box Plot - Petal Length")
plt.ylabel("Petal Length (cm)")
plt.show()

# 💡 Explanation:
# - The line inside the box = median (middle value)
# - The box = middle 50% of data
# - Dots outside = outliers
# Useful to understand spread and detect extreme values.
```

```python
# -------------------------------------------------------------
# 8. PIE CHART
# -------------------------------------------------------------
# Pie charts show parts of a whole.

species_counts = df["target"].value_counts()
plt.pie(species_counts, labels=iris.target_names, autopct="%1.1f%%", startangle=180)
plt.title("Pie Chart - Species Distribution")
plt.show()

# 💡 Explanation:
# Each slice represents how many samples belong to each species.
# Best for showing proportions or percentages.
```

```python
# -------------------------------------------------------------
# 9. SUBPLOTS
# -------------------------------------------------------------
# Subplots allow you to show multiple graphs side by side.

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# Left: Histogram
ax[0].hist(df["sepal length (cm)"], bins=15, color='orange')
ax[0].set_title("Sepal Length Distribution")

# Right: Scatter Plot
ax[1].scatter(df["petal length (cm)"], df["petal width (cm)"], color='blue')
ax[1].set_title("Petal Length vs Width")

plt.suptitle("Subplots Example - Two Graphs Together")
plt.show()

# 💡 Explanation:
# Subplots are helpful when comparing two related visualizations in one view.
```

```python
# -------------------------------------------------------------
# 10. HEATMAP (using imshow)
# -------------------------------------------------------------
# Heatmaps display numeric values using colors — great for correlation.

corr = df.corr(numeric_only=True)

plt.imshow(corr, cmap='coolwarm')
plt.colorbar(label='Correlation Strength')

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha='right')
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Heatmap - Feature Correlations")
plt.show()

# 💡 Explanation:
# Darker colors = stronger relationship between features.
# Helps understand which features are related.
```
