```python
import sys

def solve():
    data = sys.stdin.read().strip().split()
    if not data:
        return

    t = int(data[0])
    ans = []
    idx = 1

    for _ in range(t):
        x = int(data[idx])
        y = int(data[idx + 1])
        idx += 2

        if y == 2 * x:
            ans.append("NO")
        else:
            ans.append("YES")

    print("\n".join(ans))

if __name__ == "__main__":
    solve()
```
