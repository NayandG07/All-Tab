```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

data = {
    "Order_ID": ["O001","O002","O003","O004","O005","O006","O007"],
    "Category": ["Electronics","Clothing","Furniture","Electronics","Clothing","Furniture","Electronics"],
    "Region": ["North","South","West","East","North","South","West"],
    "Sales": [12000,5000,8000,15000,4000,7000,11000],
    "Profit": [3000,1200,1500,4000,800,1300,2500],
    "Discount": [10,20,15,5,25,10,12],
    "Quantity": [2,5,3,4,6,2,3],
    "Customer_Rating": [4.5,4.2,4.0,4.8,3.9,4.1,4.6]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Sales vs Profit
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Category', size='Quantity', palette='Set2')
plt.title('Sales vs Profit by Category')

```

```python
# 2. Category, Region, Sales, Profit
sns.barplot(data=df, x='Category', y='Sales', hue='Region', palette='Set1')
plt.title('Sales by Category and Region')
plt.show()
```
