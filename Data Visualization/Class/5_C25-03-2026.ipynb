```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
flowers_df = sns.load_dataset('iris')
```

```python
flowers_df
```

```python
flowers_df.species.unique()
```

```python
plt.plot(flowers_df.sepal_length, flowers_df.sepal_width, "or")
```

```python
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width)
```

```python
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width, hue=flowers_df.species)
```

```python
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width, hue=flowers_df.species, s=50)
```

```python
plt.title("Distribution of Sepal Length")
sns.histplot(flowers_df.sepal_length)
```

```python
plt.title("Distribution of Sepal width")
plt.hist(flowers_df.sepal_width)
```

```python
flowers_df.describe()
```

```python
plt.hist(flowers_df.sepal_width, bins=5)
```

```python
plt.hist(flowers_df.sepal_width, bins=np.arange(2, 5, 0.25))
```

```python
plt.hist(flowers_df.sepal_length, bins=[1, 1.5, 2, 2.5, 3, 3.5])
```

```python
setosa_df = flowers_df[flowers_df.species == "setosa"]
versicolor_df = flowers_df[flowers_df.species == "versicolor"]
virginica_df = flowers_df[flowers_df.species == "virginica"]

plt.hist(setosa_df.sepal_width, bins=np.arange(2, 5.5, 0.25), alpha=0.5, label="setosa")
plt.hist(versicolor_df.sepal_width, bins=np.arange(2, 5.5, 0.25), alpha=0.5, label="versicolor")
plt.hist(virginica_df.sepal_width, bins=np.arange(2, 5.5, 0.25), alpha=0.5, label="virginica")
plt.legend()
```

```python
plt.bar(x=flowers_df.species, height=flowers_df.sepal_length)
```

```python
# Bar Graph
sns.barplot(x=flowers_df.species, y=flowers_df.sepal_length)
```

```python
sns.barplot(x=flowers_df.species, y=flowers_df.sepal_length, data=flowers_df, hue=flowers_df.species)
```
