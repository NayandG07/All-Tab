```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("hstack:", np.hstack([a, b])) # hstack stacks arrays horizontally (column-wise)
print("vstack:\n", np.vstack([a, b])) # vstack stacks arrays vertically (row-wise)
print("dstack:\n", np.dstack([a, b])) # dstack stacks arrays depth-wise (along the third axis)
print("concatenate axis=0:", np.concatenate([a, b])) # concatenate stacks arrays along the specified axis (default is axis=0)
```

```python
A = np.arange(1, 7).reshape(2, 3)
B = np.arange(7, 13).reshape(2, 3)
print("A:\n", A)
print("B:\n", B)
print("hstack 2D:\n", np.hstack([A, B])) # hstack stacks arrays horizontally (column-wise)
print("vstack 2D:\n", np.vstack([A, B])) # vstack stacks arrays vertically (row-wise)
print("dstack 2D:\n", np.dstack([A, B])) # dstack is used to stack arrays along the third axis (depth-wise), so it will create a new dimension and stack A and B along that dimension
```

```python
M = np.arange(8).reshape(2, 4)
print(M)
print(np.split(M, 2, axis=1)) # split splits an array into multiple sub-arrays along the specified axis (default is axis=0), so here we are splitting M into 2 sub-arrays along the columns (axis=1)
```

```python
M = np.arange(8).reshape(2, 4)
print(M)
print(np.split(M, 4, axis=1))
```

```python
M = np.arange(16).reshape(4, 4)
print("M:\n", M)
print("hsplit:\n", np.hsplit(M, 4)) # hsplit splits an array into multiple sub-arrays horizontally (column-wise)
print("vsplit:\n", np.vsplit(M, 4)) # vsplit splits an array into multiple sub-arrays vertically (row-wise)
print("array_split uneven cols (3 Parts):\n", np.array_split(M, 3, axis=1)) # array_split splits an array into multiple sub-arrays along the specified axis (default is axis=0)
```

```python
x = np.array(["a", "b", "f", "a"])
print("sorted copy:", np.sort(x)) # sort returns a sorted copy of the array, so changes to the sorted array will not affect the original array and vice versa
print("argsort:", np.argsort(x)) # argsort returns the indices that would sort the array, so changes to the sorted array will not affect the original array and vice versa
```

```python
y = np.array([5, 0, 7, 0, 9, 0])
print("Where y !=0 -->", np.where(y != 0))
print("Where y ==0 -->", np.where(y == 0))
print("Nonzero -->", np.nonzero(y))
```

```python
sorted_arr = np.array([2, 4, 6, 8, 10])
print("Insert positions for [1, 5, 9]:", np.searchsorted(sorted_arr, [1, 5, 9])) # searchsorted returns the indices where the elements should be inserted to maintain order
print("Left vs Right side for equal values:")
dup = np.array([1, 2, 2, 2, 3])
print("Left side for 2:", np.searchsorted(dup, 2, side='left')) # searchsorted with side='left' returns the index of the first occurrence of the value
print("Right side for 2:", np.searchsorted(dup, 2, side='right')) # searchsorted with side='right' returns the index of the last occurrence of the value + 1
```

```python
sorted_arr = np.array([2, 4, 6, 8, 10])

# Get positions where new values should be inserted
values = [1, 5, 9]
positions = np.searchsorted(sorted_arr, values)

print("Positions:", positions)

# Insert values at the found positions
new_arr = np.insert(sorted_arr, positions, values)
print("New array after insertion:", new_arr)
```

```python
arr1 = np.array([7, 8, 6, 3, 4, 12])
val = [2, 1, 5, 9]

print("Positions:", np.searchsorted(np.sort(arr1), np.sort(val)))
print("New array after insertion:", np.insert(np.sort(arr1), np.searchsorted(np.sort(arr1), np.sort(val)), np.sort(val)))
```

```python
arr1 = np.array([7, 8, 6, 3, 4, 12])
val = [2, 1, 5, 9]

arr2 = np.sort(arr1)
print("Sorted array:", arr2)
print("Positions:", np.searchsorted(arr2, np.sort(val)))
print("New array after insertion:", np.insert(arr2, np.searchsorted(arr2, np.sort(val)), np.sort(val)))
```
