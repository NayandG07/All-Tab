### Qs. WAP that simulates a bank account. The program should ask for their initial balance and then provide option to deposit, withdraw, check balance. Use a while loop to keep the account running until the user chooses to exit.

```python
# Ask for the initial balance
balance = float(input("Enter your initial balance: "))

while True:
    # Display the menu options
    print("\nBank Account Menu:")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Check Balance")
    print("4. Exit")
    
    # Get the user's choice
    choice = int(input("Enter your choice (1/2/3/4): "))
    
    if choice == 1:
        # Deposit
        amount = float(input("Enter the amount to deposit: "))
        balance += amount
        print(f"Amount deposited: {amount}. New balance: {balance}")
    elif choice == 2:
        # Withdraw
        amount = float(input("Enter the amount to withdraw: "))
        if amount < balance:
            balance -= amount
            print(f"Amount withdrawn: {amount}. New balance: {balance}")
        else:
            print("Insufficient funds!")
    elif choice == 3:
        # Check balance
        print(f"Your current balance is: {balance}")
    elif choice == 4:
        # Exit
        print("Exiting the program. Thank you for using our services!")
        break
    else:
        print("Invalid choice! Please enter a valid option.")
```

### Qs. Greatest of four numbers from user input.

```python
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
c = int(input("Enter the third number: "))
d = int(input("Enter the fourth number: "))

greatest = max(a, b, c, d)

print("The greatest number is:", greatest)
```

# List Comprehension

```python
numbers = [1, 2, 3, 4, 5]
result = ["Even" if i % 2 == 0 else "Odd" for i in numbers]
print(result)
```

```python
num = [-5, 3, -1, 7, -9, 2]
result = [i if i > 0 else 0 for i in num]
print(result)
```

```python
## Categorise the numbers in the list as positive, negative, or zero
num = [-5, 3, -1, 0, 7, -9, 2]
result = ["Positive" if i > 0 else "Negative" if i < 0 else "Zero" for i in num]
print(result)
```
