### Qs.3 WAP in Python that calculates the factorial of a given number using while loop.

```python
num = int(input("Enter a number to find the factorial: "))
factorial = 1

i = num
while i > 0:
    factorial *= i
    i -= 1
print(f"The factorial of {num} is {factorial}")
```

### Qs.4 WAP that prints the number from 10 to 1

```python
i = 10
while i >= 1:
    print(i, end=" ")
    i -= 1
```
