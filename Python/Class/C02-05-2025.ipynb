### Difference between the modes of open built-in functions.

#### All the modes will alllow writing to the files and says that they open the files for appending and writing and updating specifically. 
#### ***1. 'r'*** - Open text file for reading. The stream is positioned at the beginning of the file.
#### ***2. 'r+'*** - Open for reading and writing. The stream is positioned at the beginning of the file.
#### ***3. 'w'*** - Truncate the file to zero length or create text file for writing. The stream is positioned at the begining of the file.
#### ***4. 'w+'*** - Open for reading and writing. The file is created if it doesn't exist, otherwise it is truncated. The stream is positioned at the beginning of the file.
#### ***5. 'a'*** - Open for writing. The file is created if it dosn't exist. The stream is positioned at the end of the file. Subsequent writes to the file will always end up at the current end of the file irrespective of any intervening position.
#### ***6. 'a+'*** - Open for reading and writing. The file is created if it doesn't exist. The stream is positioned at the end of the file. Sebsequent writes to the file will always end up at the current end of the file irrespective of any intervening postion.

```python
f = open("test.txt", "r")
f.seek(5) # Move the cursor to the 5th byte of the file and read from there
print(f.read())
f.close()
```

```python
f = open("test.txt", "a")
f.write("Jabba")
f.close()
```

#### ***"With"*** Syntax - 
- To deal with file command, with syntax is used.
- Used for safe file handling, so the file is automatically closed after the indentation block is done.

```python
with open("test.txt", "r") as f:
    print(f.read())

# The with statement automatically closes the file after the block of code is executed
```

```python
with open("test.txt", "w") as f:
    f.write("This is the new Data")
```
