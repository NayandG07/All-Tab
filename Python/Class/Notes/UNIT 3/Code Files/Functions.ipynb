```python
# Max , min, sum

def maximum(a):
    x=list(a)
    x.sort()
    return x[-1]
```

```python
q = (10,70,100,9,100,203,40000,1012,323)

print(maximum(q))
```

```python
a = '122345'
print(maximum(a))
```

```python
# Map Function

nums = [1, 2, 3, 4, 5]

def sq(n):
    return n*n

square = map(sq, nums)
print(list(square))
```

```python
# Zip Function 

name = ['Manju', 'Nikhil', 'Sambhav']
ID = [4, 1, 3]

stored = zip(name, ID)
print(stored)
```

The output <zip object at 0x0000018F11DBBB80> indicates that the zip() function in Python has returned a zip object. This object is an iterator that aggregates elements from multiple iterables (like lists or tuples). The part 0x0000018F11DBBB80 is the memory address where this zip object is stored.

```python
print(list(stored))
```

In Python, dir() is a built-in function that lists the attributes and methods of an object. When you use dir(__builtins__), it lists all the built-in functions, exceptions, and other objects that are available by default in the Python interpreter without needing to import any modules.

__builtins__ is a reference to the built-in namespace where all of Python's built-in functions (like len(), print(), int(), etc.) and exceptions (like ValueError, TypeError, etc.) are stored.
The output of this command is a list of all these built-ins, such as:
