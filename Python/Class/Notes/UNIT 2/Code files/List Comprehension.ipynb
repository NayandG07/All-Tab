```python
marks = [20, 30, 40, 50, 70]
new_m = []
for i in marks:
    i = i+2
    new_m.append(i)
print(new_m)
    
    
```

```python
marks = [20, 30, 40, 50, 70]
new_m = [i+2 for i in marks]
print(new_m)
```

```python
cubes = []
for x in range(11):
    if x%2==0:
        cubes.append(x**3)
print("Using for loop", cubes)
```

```python
cubes = [x**3 for x in range(11) if x%2 == 0]
print("Using List Comprehension", cubes)
```

```python
# Find among this list, which element is odd and whihc is even and append the result in a new list
numbers = [1, 2, 3, 4, 5, 6]
result = []

for num in numbers:
    if num % 2 == 0:
        result.append("Even")
    else:
        result.append("Odd")

print(result)
# Output: ['Odd', 'Even', 'Odd', 'Even', 'Odd', 'Even']
```

```python
numbers = [1, 2, 3, 4, 5, 6]
result = ["Even" if num % 2 == 0 else "Odd" for num in numbers]
print(result)
# Output: ['Odd', 'Even', 'Odd', 'Even', 'Odd', 'Even']
```

```python
#  Replace Negative Numbers with Zero
nums = [-5, 3, -1, 7, -9, 2]
for n in nums:
    if n>0:
        result.append(n)
    else:
        result.append(0)
print(result)
```

```python
#  Replace Negative Numbers with Zero
nums = [-5, 3, -1, 7, -9, 2]
result = []
for n in nums:
    if n>0:
        result.append(n)
    else:
        result.append(0)
print(result)
```

```python
nums = [-5, 3, -1, 7, -9, 2]
result = [num if num > 0 else 0 for num in nums]
print(result)
```

```python
# Categorize Numbers as "Positive", "Negative", or "Zero"
nums = [-3, 0, 4, -1, 9]
result = []

for num in nums:
    if num > 0:
        result.append("Positive")
    elif num < 0:
        result.append("Negative")
    else:
        result.append("Zero")

print(result)
```

```python
nums = [-3, 0, 4, -1, 9]
result = ["Positive" if num > 0 else "Negative" if num < 0 else "Zero" for num in nums]
print(result)
# Output: ['Negative', 'Zero', 'Positive', 'Negative', 'Positive']
```

```python
#Mark Passing and Failing Scores
scores = [45, 78, 30, 90, 55]
result = []
for m in scores:
    if m >= 50:
        result.append("Pass")
    else:
        result.append("Fail")
print(result)
```

```python

result = ["Pass" if score >= 50 else "Fail" for score in scores]
print(result)
# Output: ['Fail', 'Pass', 'Fail', 'Pass', 'Pass']
```

```python
# Convert Temperatures (Below 15°C as "Cold", Otherwise "Warm")
temps = [10, 22, 5, 18, 30]
result = []

for temp in temps:
    if temp< 15:
        result.append('Cold')
    else:
        result.append("Warm")
        
print(result)
```

```python
# If temp is less than 10 it is cold else 
temps = [10, 22, 5, 18, 30]
result = ["Cold" if temp < 15 else "Warm" for temp in temps]
print(result)
# Output: ['Cold', 'Warm', 'Cold', 'Warm', 'Warm']
```
