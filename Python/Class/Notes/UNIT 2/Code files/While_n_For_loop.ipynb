## While loop
- With the while loop we can execute  a set of statements as long as a condition is true


```python
num = 1

while num < 11:
    print(num)
    num += 1
```

```python
num = 0

while num < 11:
    print(num)
    if num == 3:
        break
    num+=1
```

```python
num = 0

while num < 6:
    num += 1
    if num == 3:
        continue
    print(num)
```

```python
# prime no. check
# logic is... take the input's half for range. like for finding whether 12 is prime or not. check it to 12/2 = 6.
# WAP to check whether the input by user is a prime no or not
a = int(input("Enter the no.: "))  # a =11

flag = 0

for i in range(2,(a//2+1)):
    if a%i==0:
        flag = 1
        break


if flag==0:
    print("The no. is prime ")
else:
    print("No, it is not a prime no.")
```
