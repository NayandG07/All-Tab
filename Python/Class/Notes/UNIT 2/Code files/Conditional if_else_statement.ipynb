```python
''' Write a Python program that validates a user's login by checking if the username is in uppercase, 
the password is at least 10 characters long, and the email ends with "@gmail.com". 
If all conditions are met, print a success message; otherwise, display the respective validation error.'''


u = input("Enter your username: ")
p = input("Enter your password: ")
email= input("Enter your email address : ")


if u.isupper():
    if len(p) >= 10:
        if email.endswith('@gmail.com'):
            print("You login is successful.....")
        else:
            
            print("Email address must end with @gmail.com")
    else:
        print("Length of password must be 10")
    
else:
    print("Username must be in uppercase")
```

```python
# Take input from the user
num = int(input("Enter a positive number: "))

# Initialize sum variable
sum_numbers = 0
i = 1  # Counter

# Use while loop to calculate sum
while i <= num:
    sum_numbers += i
    i += 1  # Increment counter

# Print the result
print("Sum of numbers from 1 to", num, "is:", sum_numbers)
```

```python
# Write a Python program that uses a while loop to print the numbers from 1 to 10, but skips the number 5.

num = 1
while num <= 10:
    if num == 5:
        num += 1  # Skip 5 and continue to the next number
        continue
    print(num)
    num += 1
```

```python
'''Create a Python program that asks the user for a password. If the password is incorrect, it should keep
asking until the correct password is entered. Use a while loop to achieve this.'''

correct_password = "python123"  # Set the correct password
user_password = " "

while user_password != correct_password:
    user_password = input("Enter password: ")
    if user_password != correct_password:
        print("Incorrect password, try again!")

print("Access granted!")
```

```python
# Write a Python program that calculates the factorial of a given number using a while loop.
num = int(input("Enter a number: "))
factorial = 1
i = 1

while i <= num:
    factorial *= i
    i += 1

print(f"Factorial of {num} is {factorial}")
```

```python
''' Write a Python program that simulates a bank account. The program should ask the user 
for their initial balance and then provide options to deposit, withdraw, or check their balance. 
Use a while loop to keep the program running until the user chooses to exit.'''

balance = float(input("Enter your initial balance: "))

while True:
    print("\n1. Deposit")
    print("2. Withdraw")
    print("3. Check Balance")
    print("4. Exit")
    
    choice = input("Choose an option among 1, 2, 3, 4: ")

    if choice == "1":
        amount = float(input("Enter deposit amount: "))
        balance += amount
        print(f"Amount deposited successfully! New balance: {balance}")

    elif choice == "2":
        amount = float(input("Enter withdrawal amount: "))
        if amount > balance:
            print("Insufficient balance!")
        else:
            balance -= amount
            print(f"Withdrawal successful! New balance: {balance}")

    elif choice == "3":
        print(f"Your current balance is: {balance}")

    elif choice == "4":
        print("Thank you for using our bank!")
        break  # Exit the loop

    else:
        print("Invalid choice! Please select a valid option.")
```
