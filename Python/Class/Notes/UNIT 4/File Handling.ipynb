### With Syntax

```python
with open("demo.txt", 'r') as f:  # f is the file object 
    data = f.read()
    print(data)
    
    
# to close the file is not necessary here, as 'with' will automatically close it
```

```python
#Buffered reading and writing
# it allows the program to read and write data in chunks
#a file id of 100kb = you want to read only 10kb portion of the file
```

```python
import io
with open('test_buf.txt','wb') as f:
    file = io.BufferedWriter(f)
    file.write(b"")
    file.write(b"Hello, How are you\n")
    file.write(b'''Do you know about Laitlum, Laitlum Canyons, also known as Laitlum Canyon, is a scenic and less-explored tourist destination located in the East Khasi Hills of Shillong, Meghalaya. \nThe name "Laitlum" translates to "end of hills" and the canyons offer panoramic views of the state of Meghalaya, \nwith dramatic scenery that \nshowcases the beauty of the region.''')
    file.write(b"Enjoy to the fullest")
    file.flush()   # used to write the data in chunks

# Data isn't immediately written to disk with every write().
# flush() manually clears the buffer, writing all pending data to the file.
# If you want to ensure data is physically saved before closing the file.
#Useful for large files, network streams, or periodic writes (e.g., logs).
```
