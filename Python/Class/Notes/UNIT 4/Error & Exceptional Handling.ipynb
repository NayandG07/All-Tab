```python
def validate_salary(salary):
    if salary <= 0:
        print("Negative salary is not possible")
        
    elif salary >= 30000000:
        print("Salary is unexpected")

    else:
        print("Salary is valid")
```

- Now lets make a custom exeption class based on the logic for validated salary

```python
class validateSalary(Exception):  # exception is the parent/ base class where python exp are avl
    def __init__(self, msg):
        self.msg = msg
```

```python
# instead of print use raise keyword
def validate_salary(salary):
    if salary <= 0:
        raise validateSalary("Negative salary is not possible")
        
    elif salary >= 30000000:
        raise validateSalary("Salary is unexpected")

    else:
        raise validateSalary("Salary is valid")
```

```python
try:
    salary = int(input("Enter your salary"))
    validate_salary(salary)
except validateSalary as e:
    print(e)
```

```python
'''Write a program that tries to open a file named data.txt. 
If the file does not exist, catch the exception and print 
a user-friendly message.'''
```

```python
try:
    with open('data.txt', 'r') as file:
        content = file.read()
        print(content)
except FileNotFoundError:
    print("File not found. Please check the file name.")
```

```python
'''Write a program that asks the user to enter an integer. If the input is not an integer,
handle the error and ask the user to enter it again, until they enter a valid integer.'''
```

```python
while True:
    try:
        num = int(input("Enter a number: "))
        print("You entered:", num)
        break  # Exit the loop if input is valid
    except ValueError:
        print("Invalid input! Please enter a number.")
```

```python
'''Create a custom exception called InvalidAgeError. Write a program that 
accepts an age input and raises InvalidAgeError if the age is less than 18.'''
```

```python
class InvalidAgeError(Exception):
    pass

try:
    age = int(input("Enter your age: "))
    if age < 18:
        raise InvalidAgeError("Age must be 18 or older.")
    else:
        print("Access granted.")
except InvalidAgeError as e:
    print("InvalidAgeError:", e)
except ValueError:
    print("Please enter a valid integer for age.")
```
