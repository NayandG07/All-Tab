```python
a = ["rohan", 100, 300, "raju"]
# Slicing
print(a[1:3:1])
#inner element indexing
print(a[0][2])  
```

```python
a=[100,200,[10,"good","###morning###",[30,40,"every"],50],"one",60]

print(a[2][3])  # It will print the innermost list

print(a[2][3]+a)   # it will add and print the innermost and the main list

print ###morning### using indexing
print(a[2][2])

# print "eve" from the word every (indexing + sicing)
print(a[2][3][2][0:3:1])

# print good morning everyone
print(a[2][1] + " "+a[2][2][3:10:1] + " "+a[2][3][2] +  a[3])

```

```python
a = [10, 20, 30]
a.append("rohan")
a.append(1)
a.append("kumar")
print(a)
```

```python
a = [10,20,30]

a.insert(1,'raju')
# insert(index no., the new element)
print(a)
```

```python
a = [1,2,3]
print(a, id(a))
b = [4,5,6]
print(id(b))
a.extend(b)

print(a)   # it will assign at the same id of a 
print(id(a))
```

```python
# whereas if we use concatenation
#it will assign at a different location id of a

a= [1,2,3,4,5]
print(a, id(a))
b=[10,20,30]

a = a+b
print(a)
id(a)
```

```python
a = [10,20,30,40,50]

a.pop()
print(a)
print(a.pop())

a.pop(2)
print(a)
```

```python
a = [10,20,30,40,50]
a.remove(30)  # it remove with the parameter element
print(a)
```

```python
a = [10,20,30,40,50]
print(a.clear())

a.clear()
print(a)
```

```python
## count()

a = [10,30,10,30,20,10,50,10,30,20]
print(a.count(10))
```

```python
a = [10,30,10,30,20,10,50,10,30,20]
print(a.index(20))
print(a.index(10,3))   # index(element to be searched, the no. of times it occured)
```

```python
a = [10,30,10,30,20,10,50,10,30,20]

a.sort()
print(a)
```

```python
a = [10,30,10,30,20,10,50,10,30,20]
a.sort(reverse= True) # for descending order
print(a) 
```

```python
a = [10,30,10,30,20,10,50,10,30,20]
a.sort(reverse= False) # for ascending order
print(a) 
```

```python
l = ["zth",'ac','egtht','cefs']

l.sort()
print(l)
```
