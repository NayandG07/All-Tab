### PYTHON OPERATORS
- Arithmetic Operators
- Assignment Operators
- Comparision Operators
- Logical Operators
- Membership Operators

```python
# Arithmetic Operators (+,-,*,/,%,**) # used with numerical values to perform common mathematics calculation
# eg , print(x+y, x-y, x*y........)
x = 2
y = 3
print(x+y)
print(x**y)
```

```python
# Assignment Operators (=, +=, -=, *=, /=, %=, **=)   # Used to Assign vaues to variabes
x, y = 5, 9
x = y
print(x)
```

```python
x, y = 100, 9
x += 10 # x = x+y
print(x)

```

```python
x *= y # x = x * y
print(x)
```

```python
# Division are of 3 types
# / ==> Single slash  = > Gives quotient value in float form
# // ==> Double slash ==> gives integer form 
# % ==> Modulus ==> R => integer
```

```python

print(7%50)
```

```python
# Comparison Operators (==, !=, <, <=, >=) # Used to compare two values
# print(x == y, x!=y, x<y, x>y, x<=y)

x,y = 10, 41

x<=y
```

```python
# logical operators (and, or, not)  # Used to combine conditional statement
# print(x and y, x or y, not y)

x,y = 10, 41

print((x < y) and (x==y))
print(not((x < y) and (x==y)))
print((x < y) or (x>=y))
```

```python
# Membership Operators (in ,not in) # Used to test if  sequence is presented in an object
# print(x in y, x not in y)
# in checks for existence.
# not in checks for non-existence.
```

```python
print('a' in 'apple')          # True (because 'a' exists in 'apple')
print('z' in 'apple')          # False (because 'z' is not in 'apple')

# Using 'not in'
print('x' not in 'python')     # True (because 'x' is not in 'python')
print('py' in 'python')        # True (substring 'py' exists in 'python')
```
