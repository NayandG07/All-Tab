## Dictionary


- Dictionaries are used to store data values in key:value pairs
- A dictionary is a collection which is ordered*, changeable and do not allow duplicates

```python
## DICTIONARY

a = {10 : "raju", 20 : "amit", "vipul" :3000}
print(a)

print(type(a))
print(len(a))
```

```python
a.get(20)
```

```python
a = {}
print(type(a))
```

```python
a = {}
a=set(a)
print(type(a))
```

```python
a = {10:"rohan", 30:"raju", (1,2,3,4):2000, 10:"mohit"}
print(a)  # dont take multiple ids which is duplicate
```

```python
a = {10:"rohan", 30:"raju", "vipul":2000}
print(a[30])
print(a[30][2])
```

```python
# to add any new pair
a = {10:"rohan", 30:"raju", "vipul":2000}

#dict_name[key] = value   to add a new element

a[20] = "utkarsh"
print(a)
```

```python
a = {10:"rohan", 30:"raju", "vipul":2000}

b =list(a.keys())

print(b,type(b))
```

```python
a = {10:"rohan", 30:"raju", "vipul":2000}

b = a.keys()
print(b)
```

```python
a = {10:"rohan", 30:"raju", "vipul":2000}

b = a.values()
print(b)
```

```python
a = {10:"rohan", 30:"raju", "vipul":2000}

print(a.items())
```

```python
x = [10,30,"vipul"]
y = ['rohan', 'raju', 2000]

d = {}

d[x[0]]=y[0]
d[x[1]]=y[1]
d[x[2]]=y[2]


print(d)   # this could also be done using zip()
```

```python
a = ['Punjab','Assam','Rajasthan','AP']
b = ['Bhangra','Bihu','Ghoomar','Santhali']

d = {}
d[a[0]]=b[0]
d[a[1]]=b[1]
d[a[2]]=b[2]
d[a[3]]=b[3]
print(d)
```

```python
x = [10,30,"vipul"]
y = ['rohan', 'raju']

a = dict(zip(x,y))
print(a)
```

```python
x = [10,30,"vipul"]
y = ['rohan', 'raju', 2000]

a = tuple(zip(x,y))
print(a)
```

```python
a = {10: "rohan", 30 :"raju", "vipul" : 2000}

a.pop(30)
print(a)
```

```python
a = {10: "rohan", 30 :"raju", "vipul" : 2000}

a.update({20:"utkarsh", 40 : "yash"})
print(a)
```

```python
a = {10: "rohan", 30 :"raju", "vipul" : 2000}
b = {20:"utkarsh", 40 : "yash"}
 
a.update(b)
print(a)
```
