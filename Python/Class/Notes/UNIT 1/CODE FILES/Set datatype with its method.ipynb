```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7}

# union
print(s1.union(s2))
print(s1 | s2)

# intersection

print(s1.intersection(s2))
print(s1-s2)
print(s2-s1)

# symmetric_difference
print(s1.symmetric_difference(s2))
print(s1^s2)
```
