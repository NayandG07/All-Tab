```python
values = tuple(input("Enter comma-separated values: ").split(","))
print(f"Your tuple: {values}. The data type is {type(values)}.")
```

```python
keys = input("Enter keys (comma-separated): ").split(",")
values = input("Enter values (comma-separated): ").split(",")
dictionary = dict(zip(keys, values))
print(f"Your dictionary: {dictionary}. The data type is {type(dictionary)}.")
```

```python
#Q4: What will be the output of the following code?

num = 255
print(f"Decimal: {num}, Hex: {num:x}, Octal: {num:o}, Binary: {num:b}")
```

```python
# Right-Align Text Formatting
#Q5: What will be the output of the following code?
print("{:>10}".format("Python"))
```

```python
#. Left-Align and Center-Align
#Q6: Fill in the blanks to correctly format the text:

text = "Hello"
print("{:<10}".format(text))  # Left-align
print("{:^10}".format(text))  # Center-align
```

```python
 #Using Format Specifiers with Operators
#Q7: What will be the output of this program?

a = 8
b = 3
print(f"{a} divided by {b} is {a/b:.2f}")
```

```python
#Formatting Boolean Values
#Q9: What will be the output of this program?

value = True
print(f"Boolean value: {value:d}")
```

```python
#Combining Different Format Specifiers
#Q10: What will be the output of this program?
a, b = 5, 2
print(f"Sum: {a + b}, Difference: {a - b}, Product: {a * b}, Division: {a / b:.1f}")
```
