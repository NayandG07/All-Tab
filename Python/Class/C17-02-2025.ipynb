### Apply append() to add "Orange" and use sort(),reverse(),remove(banana),pop(1), insert(1, "Watermelon"), search index("cherry")

```python
fruits = ["apple", "banana", "cherry"]
fruits.append("Orange")
print(fruits)
fruits.sort()

fruits.reverse()
fruits.remove("banana")
print(fruits)

fruits.pop(1)

fruits.insert(1, "watermelon")
print(fruits)
print(fruits.index("cherry"))
```

# Tuples
- ### Tuples are used to store multiple /items into a single variable which is ordered and unmutable.
- ### It is written with () [Ex- Tup(1,2,3)]
- ### Allows duplicate element

```python
num = (1,2,4,5,1,2,5,7,10,1,1,4)
print(num.count(1))
print(num.index(7))
print(min(num)) # prints the smallest number in the list
print(max(num)) # prints the largest number in the list
```

## Append 6000 in between (5000,7000)
### a = (10,20,30,(300,400,(5000,7000),500),400)

```python
# Define the nested tuple
a = (10, 20, 30, (300, 400, (5000, 7000), 500), 400)

# Converted the nested tuple to a list, modified it, and converted it back to a tuple
a_list = list(a)
a_list[3] = list(a_list[3])
a_list[3][2] = list(a_list[3][2])

a_list[3][2].insert(1, 6000) # Insert 6000 at index 1 in the innermost list

a_list[3][2] = tuple(a_list[3][2])
a_list[3] = tuple(a_list[3])
A = tuple(a_list)

print(A)
```
