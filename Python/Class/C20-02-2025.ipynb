# Sets
- #### Sets are used to store multiple items in a single variable.
- #### Sets are written with {} 
- #### This is the best data type to remove duplicates.
- #### It always gives random values.
- #### Mutable.

```python
s1 = {100,1000,10,10,20}
s1.add(20000) 
s1.remove(10)
print(s1)

# s1.remove(500)
s1.discard(500) # No error occurs if value is not valid
print(s1)
s1.pop() # Delete any Random element
```

```python
S1 = {1,2,3,4,5}
S2 = {4,5,6,7}

print(S1.union(S2))
print(S1.intersection(S2))
print(S1.difference(S2))
print(S1.symmetric_difference(S2))
print(S1^S2)
```

#### frozenset() ---> Becomes immutable.

```python
a = frozenset(S1)
print(type(a))
```

```python
fruits = {'apple', 'cherry', 'banana'}

# fruits.discard('banana') # can be used to avoid key-errors
fruits.remove('banana')
print(fruits)
```

```python
x = {'Apple', 'cherry', 'banana'}
y = {'Google', 'Microsoft', 'Apple'}

print(x.union(y))
print(x.difference(y))
print(x.intersection(y))
x.update(y) 
print(x)
```

### Qs.1 Write a dictionary with Hindi words with values as their English translations. Provide user with an option to look it up or check it.

```python
# Dictionary with Hindi words and their English translations
hindi_to_english = {
    "Namaste": "Hello",
    "Dhanyavad": "Thank you",
    "Paani": "Water",
}

while True:
    print("\nOptions:")
    print("1. Look up a word")
    print("2. Exit")
    choice = input("Enter your choice (1 or 2): ")
    
    if choice == '1':
        hindi_word = input("Enter a Hindi word to look up its English translation: ")
        translation = hindi_to_english.get(hindi_word)
        if translation:
            print("The English translation of '",hindi_word,"' is '",translation,"'.")
        else:
            print("The word '",hindi_word,"' is not in the dictionary.")
    elif choice == '2':
        print("Exiting the program.")
        break
    else:
        print("Invalid choice. Please enter 1 or 2.")
```

### Qs.2 WAP to input 8 numbers from user and display all the unique numbers.

### Qs.5 Create an empty dictionary to allow 4 friends to enter their favourite languages as values and use keys as their names. Assuming the names are unique.

```python
languages = {}

for i in range(4):
    name = input("Enter your name: ")
    language = input("Enter your favorite language: ")
    languages[name] = language # names are stored as keys 
                               # dict[key] = value

print("Favorite languages:", languages)
```

### Qs.6 If names of two friends are same, then what will be the problem in the prevous question?

- ##### If the names of two friends are the same, the second entry will overwrite the first entry in the dictionary. This is because dictionary keys must be unique, and if a key already exists in the dictionary, assigning a new value to that key will replace the existing value.

```python
languages = {}

for i in range(4):
    name = input("Enter your name: ")
    language = input("Enter your favorite language: ")
    languages[name] = language # names are stored as keys

print("Favorite languages:", languages)
```
