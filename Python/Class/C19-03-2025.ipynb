## Functions

```python
def add(a,b):
    return a+b

def sub(a,b):
    return a-b

def mul(a,b):
    return a*b

def div(a,b):
    return a/b

a = 7
b = 18

print(add(a,b))
print(add(10,20))

print(sub(a,b))
print(sub(10,20))

print(mul(a,b))
print(mul(10,20))

print(div(a,b))
print(div(10,20))
```

### Qs. WAP to check whether a given number is even or odd and then print the result.

```python
def num(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

a = int(input("Enter a n: "))
print(num(a))
```
