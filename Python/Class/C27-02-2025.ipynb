#### Nested if-else

```python
n = int(input("Enter number: "))
if(n<=15):
    if(n==10):
        print("Play Cricket")
    else:
        print("Play Kabadi")
else:
    print("Don't Play")
```

### Qs.1 WAP that prompts the user to enter a number and then print it's interval.

```python
a = int(input("Enter any number between 0 to 30: "))
if a >= 0 and a <= 30:
    print("It's in the range of 0-30")
    if a >= 0 and a < 10:
        print("Range is in 0-10")
    else:
        print("Range is not in 0-10")
else:
    print("Number is not between 0 to 30")
```

## Control Statements : Loops

```python
i = 1
while i < 6:
    print(i)
    i += 1
else:
    print("Loop finished")
```

```python
i = 0
while i <= 10:
    print(i, end=" ") # end =" " is used to print all the numbers in the same line
    i += 1
else:
    print("Loop finished")
```

```python
i = 0
sum = 0
while i <= 10:
    sum += i
    i += 1

print("Sum of the first 10 numbers: ",sum)
```

```python
num = [6,5,3,8,4,2,5,4]
sum1 = 0
for i in num:
    sum1 += i
print(sum1)
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    print(x)
```

```python
genre = ['pop', 'rock', 'jazz']
for i in range(len(genre)):
    print(genre[i])
```

```python
for i in range(6):
    print(i, end=(" "))
```

```python
adj = ["red", "tasty"]
fruits = ["apple", "cherry"]

for i in adj:
    for j in fruits:
        print(i, j)
```
