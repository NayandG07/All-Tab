```python
'''Create a class and perform Inhheritance'''
# Base class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f"{self.name} makes a sound."

# Derived class
class Dog(Animal):
    def speak(self):
        return f"{self.name} says Woof!"

# Example usage
dog = Dog("Buddy")
print(dog.speak())
```

```python
'''Write a Python program to demonstrate dictionary creation, value access, and printing keys and values.'''
# Dictionary creation
student = {
    "name": "Alice",
    "age": 21,
    "course": "Computer Science"
}

# Accessing values
print("Name:", student["name"])
print("Age:", student["age"])

# Printing keys and values4
print("All keys:")
for key in student.keys():
    print(key)

print("All values:")
for value in student.values():
    print(value)

print("All key-value pairs:")
for key, value in student.items():
    print(f"{key}: {value}")
```

```python
'''Write a Python function that checks if a given number is prime or not from 1 to 200.

'''
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

# Print prime numbers from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num, end=' ')
```

```python
'''Create a Python program that uses `filter()` to remove all the vowels from a given string.

'''
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda x: x not in vowels, s))
# Example usage
input_string = "Hello, World!"
result = remove_vowels(input_string)
print("Original String:", input_string)
print("String without vowels:", result)
```

```python
'''Write a Python program that returns a list of 2-tuples. Each tuple should contain the item code and the total value of the item (i.e., quantity × price per unit).
If the total value is less than 100.00 €, a restocking fee of 5.00 € should be added to the total value.

Write the Python program using lambda and map()'''
''' data :   
[  
 [101, "Wireless Mouse", 10, 12.50],  
 [102, "Mechanical Keyboard", 4, 45.00],  
 [103, "USB-C Hub", 7, 15.00],  
 [104, "Webcam HD", 2, 65.00]  
]
'''
# Data
data = [  
    [101, "Wireless Mouse", 10, 12.50],  
    [102, "Mechanical Keyboard", 4, 45.00],  
    [103, "USB-C Hub", 7, 15.00],  
    [104, "Webcam HD", 2, 65.00]  
]

# Using lambda and map to calculate total value with restocking fee if applicable
result = list(map(lambda item: (
    item[0],  # Item code
    item[2] * item[3] + (5.00 if item[2] * item[3] < 100.00 else 0)  # Total value with restocking fee
), data))

# Output the result
print(result)
```

```python
'''Create a base class Shape and derive classes Circle, Rectangle, and Triangle. 

Implement methods to calculate area and perimeter using polymorphism and inheritance.'''

# Base class
class Shape:
    def area(self):
        pass

    def perimeter(self):
        pass

# Derived class: Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius  # Using 3.14 as an approximation for π

    def perimeter(self):
        return 2 * 3.14 * self.radius

# Derived class: Rectangle
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

# Derived class: Triangle
class Triangle(Shape):
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def area(self):
        return (self.a * self.b) / 2

    def perimeter(self):
        return self.a + self.b + self.c

# Example usage
shapes = [
    Circle(5),          # Circle with radius 5
    Rectangle(4, 6),    # Rectangle with length 4 and width 6
    Triangle(3, 4, 5)   # Triangle with sides 3, 4, and 5
]

for i in shapes:
    print(f"{i.__class__.__name__}:")
    print(f"  Area: {i.area()}")
    print(f"  Perimeter: {i.perimeter()}")
```

```python
'''Write a class Fraction with attributes numerator and denominator. Override the str method to display the 
fraction as "numerator/denominator"'''
class Fraction:
    def __init__(self, numerator, denominator):
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"
    
# Example usage
fraction = Fraction(3, 4)
print("Fraction:", fraction)  # Output: Fraction: 3/4
```

```python
'''Create a class Product with attributes name, price, and quantity. Implement a method total_price() that 
calculates the total price of the product.'''
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_price(self):
        return self.price * self.quantity
# Example usage
product = Product("Laptop", 1000, 2)
print(f"Total price of {product.name}: €{product.total_price()}")  # Output: Total price of Laptop: €2000
```

```python
'''Create a class MathOperations with a class method add_numbers() to add two numbers and a static
method subtract_numbers() to subtract two numbers.

'''
class MathOperations:
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b
# Example usage
result_add = MathOperations.add_numbers(10, 5)
result_subtract = MathOperations.subtract_numbers(10, 5)
print("Addition Result:", result_add)        # Output: Addition Result: 15
print("Subtraction Result:", result_subtract)  # Output: Subtraction Result: 5
```

```python
'''Create a class House with attributes address and price. Create a derived class Mansion that adds an
attribute number_of_rooms.
'''
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

    def __str__(self):
        return f"House at {self.address} costs €{self.price}"
class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

    def __str__(self):
        return f"Mansion at {self.address} with {self.number_of_rooms} rooms costs €{self.price}"
# Example usage
mansion = Mansion("123 Luxury St", 500000, 10)
print(mansion)  # Output: Mansion at 123 Luxury St with 10 rooms costs €500000
```

```python
'''Create a base class Employee with a method calculate_salary().
Derive FullTimeEmployee, PartTimeEmployee, and Intern classes with their own salary logic.
Use inheritance and polymorphism to calculate salaries for different employees in a list.'''

class Employee:
    def calculate_salary(self):
        pass
class FullTimeEmployee(Employee):
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def calculate_salary(self):
        return self.salary
class PartTimeEmployee(Employee):
    def __init__(self, name, hourly_rate, hours_worked):
        self.name = name
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked

    def calculate_salary(self):
        return self.hourly_rate * self.hours_worked
class Intern(Employee):
    def __init__(self, name, stipend):
        self.name = name
        self.stipend = stipend

    def calculate_salary(self):
        return self.stipend
# Example usage
part = PartTimeEmployee("Alice", 20, 15)  # $20/hour for 15 hours
full_time = FullTimeEmployee("Bob", 3000)  # $3000 monthly salary
intern = Intern("Charlie", 500)  # $500 stipend

print(f"{part.__class__.__name__}'s salary: ${part.calculate_salary()}")
print(f"{full_time.__class__.__name__}'s salary: ${full_time.calculate_salary()}")
print(f"{intern.__class__.__name__}'s salary: ${intern.calculate_salary()}")
```
