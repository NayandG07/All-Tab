#### int --> float
#### float --> string(str)

```python
a = 12.5
print(type(a))
b = str(a)
print(type(b))
c = 12.8
print(c)
print(int(c))
```

```python
a = "7"
print(type(a))
b = int(a)
print(type(b))
```

```python
a = 3.14
b = complex(a)
print(b)
print(type(b))
c = 3+2j
print(c)
```
