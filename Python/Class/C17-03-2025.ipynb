### Qs. WAP to find geatest of four numbers entered by the user in a list using for loop.

```python
a = []
for i in range(1,5):
    n = int(input(f"Enter num {i}:"))
    a.append(n)

a.sort()
print(f"Max: {a[-1]}")

# Without sorting
max = 0
for i in a:
    if i > max:
        max = i
print(f"Max: {max}")

# print(f"Max: {max(a)}")
```

```python
n = int(input("Enter a number:"))
a,b = 0,1
for i in range(n):
    print(a,end=" ")
    # a,b = b,a+b
    c = a+b
    a = b
    b = c
print(f"\nFibonacci number at {n} is {a}")
```

### WAP to check whether the input by user is a prime number or not?

```python
n = int(input("Enter a number:"))

if n <= 1:
    is_prime = False
else:
    is_prime = True
    for i in range(2, n):
        if  n % i == 0:
            is_prime = False
            break
if is_prime:
    print(f"{n} is a prime number")
else:
    print(f"{n} is not a prime number")
```
