```python
print("Hello \n World") # \n is for new line --> Escape sequence
print("Hello \t World") # \t is for tab space --> Escape sequence
print("Hello \\ World") # \\ is for backslash --> Escape sequence
print("Hello \" World") # \" is for double quote --> Escape sequence
print("Hello\rWorld") # \r is for carriage return --> Escape sequence
```
