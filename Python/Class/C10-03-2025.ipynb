### Break and Continue statement
- ### Break is used to break the loop
- ### Continue is used to skip a iteration value

```python
i = 1
while i<= 10:
    print(i, end=' ')
    if i== 5:
        break
    i += 1
print('\nEnd')
```

```python
for i in range(1,11):
    print(i)
    if i == 5:
        print("Breaking the loop at", i)
        break
print("End of loop")
```

```python
i = 0
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)
```

```python
for i in range(1,6):
    if i == 3:
        continue
    print(i)
print("End of loop")
```

### Pass Statement
- The pass statement in Python is a null operation; it is used as a placeholder in situations where a statement is syntactically required but you do not want any code to be executed. It is often used in loops, functions, classes, or conditionals where you plan to add code later but want to maintain the correct syntax.
- Key Points about pass Statement
- 1.Functionality: The pass statement does nothing and is used as a placeholder.
- 2.Usage: It is used in places where a statement is syntactically required but no action is needed.
- 3.Common Use Case: It is commonly used in loops, functions, classes, or conditionals during development to indicate that code will be added later.
- 4.Effect on Execution: The pass statement has no effect on the execution of the program.
- 5.Syntax: The pass statement is written as pass.

#### Qs. WAP to calculate electricity bill 
#### Eelectricity unit by user
##### 200<= Free, 300<= 5 Rs/Reading, 400<= 10Rs/Reading, 500<= 20Rs/Reading

```python
units = int(input("Enter the Reading: "))
bill = 0

if units <= 200:
    bill = 0
elif units <= 300:
    bill = (units - 200) * 5
elif units <= 400:
    bill = (300 - 200) * 5 + (units - 300) * 10
elif units <= 500:
    bill = (300 - 200) * 5 + (400 - 300) * 10 + (units - 400) * 20
else:
    bill = (300 - 200) * 5 + (400 - 300) * 10 + (500 - 400) * 20 + (units - 500) * 30

print("Electricity Bill: ", bill)
```

#### Qs. WAP that validates a user's login by 1. Username (shoudl be in uppercase), 2. Password (shoudl be more or equal to 10 characters), 3. Email (@gmail.com is allowed only). If all conditions met, Print Login succesful, otherwise Validation error

```python
username = input("Enter your username (uppercase): ")
password = input("Enter your password (at least 10 characters): ")
email = input("Enter your email (@gmail.com only): ")

if username.isupper() and len(password) >= 10 and email.endswith("@gmail.com"):
    print("Login Succesful")
else:
    print("Validation Error")
```

```python
username = input("Enter your username (uppercase): ")
password = input("Enter your password (at least 10 characters): ")
email = input("Enter your email (@gmail.com only): ")

if username.isupper():
    if len(password) >= 10:
        if email.endswith("@gmail.com"):
            print("Login Succesful")
        else:
            print("Email Error")
    else:
        print("Password Error")
else:
    print("Username Error")
```

### Qs. Stone, Paper, Scissor Game

```python
moves = ["rock", "paper", "scissors"]
player1 = input("Enter your move: ").lower()
player2 = input("Enter your move: ").lower()

if player1 not in moves or player2 not in moves:
    print("Invalid input! Please enter rock, paper, or scissors.")
else:
    if player1 == player2:
        print("It's a tie!")
    elif (player1 == "rock" and player2 == "scissors") or (player1 == "scissors" and player2 == "paper") or (player1 == "paper" and player2 == "rock"):
        print("Player 1 wins!")
    else:
        print("Player 2 wins!")
```
