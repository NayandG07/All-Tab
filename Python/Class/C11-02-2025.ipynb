## 2. Find hypotenuse using PnB from user.

```python
b = float(input("Base of Triangle: "))
p = float(input("Height of Triangle: "))
print("hypotenuse of Triangle: ", (b**2 + p**2)**0.5)
```

## 3. Find the volume of cone.

```python
r = float(input("Radius of Cone: "))
h = float(input("Height of Cone: "))
print("Volume of Cone: ", (1/3) * 3.14 * (r**2) * h) # volume = 1/3 * pi * r^2 * h
```
