### __init__
- Purpose: The __init__ method initializes the attributes of the class.
- Self Parameter: The first parameter, self, refers to the instance being created.
- Automatic Call: It is automatically invoked when an object is instantiated (e.g., car1 = Car(...)).

```python
class Car:
    def __init__(self,brand,model,year):
        self.brand = brand
        self.model = model
        self.year = year

    def display(self):
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Year: {self.year}")

car1 = Car("Toyota", "Camry", 2020)
car1.display()
```

### Create Student class that takes name and marks of 3 subjects as arguments in cosntructor, then create a method print the average.

```python
class Student:
    def __init__(self, name, marks):
        self.marks = marks
        self.name = name

    def average(self):
        return sum(self.marks) / len(self.marks)

s1 = Student("John", [99, 98, 97])
print(s1.name, s1.marks)
print(f"Average of marks: {s1.average()}")
s1.name = "Harry"
print(s1.name, s1.marks)
```

```python
class Student:
    def __init__(self, name, marks):
        self.marks = marks
        self.name = name

    def average(self):
        sum = 0
        for i in self.marks:
            sum += i
        return sum / len(self.marks)

s1 = Student("John", [99, 98, 97])
print(s1.name, s1.marks)
print(f"Average of marks: {s1.average()}")
```

#### @staticmethod: Defines a method that doesn't depend on the instance or class.
#### @classmethod: Defines a method that takes the class (cls) as its first argument.

```python
class A:
    @staticmethod # Decorator
    def hello():
        return "Hello, I am a static method!"

a1 = A()
print(a1.hello())
```

### Create an account class with 2 attributes, Balance and account number. Create methods for debit, credit and printing the balance.

```python
class Account:
    def __init__(self, acc_num, balance=0):
        self.acc_num = acc_num
        self.balance = balance

    def credit(self, amount):
        self.balance += amount
        print(f"Credited {amount}. New balance: {self.balance}")

    def debit(self, amount):
        if amount > self.balance:
            print("Insufficient balance!")
        else:
            self.balance -= amount
            print(f"Debited {amount}. New balance: {self.balance}")

    def display(self):
        print(f"Account {self.acc_num} balance: {self.balance}")


# Example usage
acc1 = Account("123456789", 1000)
acc1.display()
acc1.credit(500)
acc1.debit(300)
acc1.debit(1500)
acc1.display()
```
