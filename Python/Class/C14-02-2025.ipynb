## Lists and it's methods
- #### Lists are used to store multiple items in a single varibale.
- #### It is a collection which is ordered, changeable(mutable) and allows duplicate members.

```python
l1 = ['a', 'b', 'c', 'd']
print(type(l1))
print(l1, type(l1), len(l1))
```

#### [start:end:step]

```python
print(l1[1:3:1]) # starts from 1, ends at 3, but skips 1
```

#### Inner element slicing

```python
l2 = ["Australia", "Japan", "Bern"]
print(l2[0][5]) # inner element indexing. prints 'a'(index = 5) from 'Australia'
```

```python
l3 = ["abc", "pul", "sbe", "yzxw"]
print(l3[2][1])
```

```python
a = [100, 200,[10, "good", "###Morning###", [30, 40, "every"], 50], "one", 60]
a[2][3]
```

```python
print(a[2][3]+a)
```

```python
# print "eve" from "every"
print(a[2][3][2][0:3])
```

```python
# print "good morning everyone"
print(a[2][1]+" "+a[2][2][3:10]+" "+a[2][3][2]+a[3])
#     good        Morning           every      one
```

### Qs.1 Check if the string "Hello123" is purely alphabatic or alpha numeric.

```python
s = "Hello123"
print(s.isalnum())
print(s.isalpha())
```

### Qs.2 Use replace and swapcase method in the string "Hello World" to replace all the spaces with underscore. 

```python
s = "Hello World!"
s1 = s.replace(" ", "_")
s1.swapcase()
```

```python
s.replace(" ", "_").swapcase()
```

### Qs.3 Split the sentence "apples, banana, cherry" by commas and then join the list back into a single string separated by semicolons (;) 

```python
s = "apples, banana, cherry"
print(s.split(", "))
print("; ".join(s.split(", ")))
```

### Qs.4 Find the position of sub string "Python" from the string "Welcome to Python" using index() and find() methods.

```python
s = "Welcome to Python"
print(s.find("Python"))
print(s.index("Python"))
```

### Qs.5 Check if the string "example.py" starts with "ex" and endswith ".py"

```python
s = "example.py"
print(s.startswith("ex"))
print(s.endswith(".py"))
```
