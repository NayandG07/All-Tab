### 3 Types of Division 
    1. / --> Single slash div. ---> float
    2. // --> Double slash div. ---> int
    3. % --> Module slash div. ---> float

```python
# Arithmetic operators
a = 7
b = 1.8

x = 8
y = 2
print(a//b) # int division
print(a/b) # float division
print(a%b) # remainder
print(a**b) # power
print(a*b) # multiplication

print(x//y) # int division
print(x/y) # float division
print(x%y) # remainder
print(x**y) # power
print(x*y) # multiplication
```
