### 1. Difference between mutable and immutable data types in python? Porvide examples

##### Mutable Data Types
- Mutable data types are those whose values can be changed in place after they are created. This means we can modify the contents of the object without creating a new object. Examples- List, Dictionary, Set, User-defined objects.

##### Immutable Data Types
- Immutable data types are those whose values cannot be changed after they are created. Any modification to an immutable object results in the creation of a new object. Examples- Tuple, String, Integer, Float, Boolean, Complex numbers.

```python
# Example of mutable data types

# List
list1 = [1, 2, 3]
list1[0] = 10
print(list1)  # Output: [10, 2, 3]

# Dictionary
dict = {"name": "Harry", "age": 25}
dict["age"] = 26
print(dict)  # Output: {'name': 'Harry', 'age': 26}

# Sets
set1 = {1, 2, 3}
set1.add(4)
print(set1)  # Output: {1, 2, 3, 4}
```

```python
# Example of immutable data types

# String
str1 = "hello"
new_string = str1.replace("h", "H")
print(str1)  # Output: hello
print(new_string)  # Output: Hello

# Tuple
tup1 = (1, 2, 3)
new_tuple = (10, 2, 3)
print(tup1)  # Output: (1, 2, 3)
print(new_tuple)  # Output: (10, 2, 3)
```

### 2. How does Python handle memory allocation for variables? Explain with reference to id() .

#### Memory Allocation for Variables in Python
- Python handles memory allocation for variables using a combination of a private heap space and a memory manager. How Python manages memory:

- Private Heap Space: All Python objects and data structures are stored in a private heap. This heap is managed by the Python memory manager, which handles the allocation and deallocation of memory for Python objects.

- Memory Manager: The memory manager is responsible for managing the private heap. It includes several components, such as the garbage collector, which automatically deallocates memory that is no longer in use.

- Reference Counting: Python uses reference counting to keep track of the number of references to each object in memory. When an object's reference count drops to zero, the memory occupied by the object is deallocated.

- Garbage Collection: In addition to reference counting, Python has a garbage collector that can detect and clean up circular references (objects that reference each other).

- Using id() to Understand Memory Allocation
The id() function in Python returns the unique identifier (memory address) of an object. This identifier is unique for the object during its lifetime.

#### Summary: 
- Memory Allocation: Python uses a private heap space managed by a memory manager to allocate memory for variables. The memory manager includes components like the garbage collector and reference counting.
- Reference Counting: Python keeps track of the number of references to each object. When the reference count drops to zero, the memory is deallocated.
- Garbage Collection: Python's garbage collector handles circular references and cleans up unused memory.
- id() Function: The id() function returns the unique identifier (memory address) of an object. It helps understand how variables reference objects in memory.

```python
# Example with integers
a = 10
b = a # Reference to the same memory address
print(id(a))  # Output: Memory address of the integer object 10
print(id(b))  # Output: Same memory address as a

# Example with lists
list1 = [1, 2, 3]
list2 = list1
print(id(list1))  # Output: Memory address of the list object
print(id(list2))  # Output: Same memory address as list1

# Modifying list2
list2.append(4)
print(list1)  # Output: [1, 2, 3, 4]
print(id(list1))  # Output: Same memory address as before
print(id(list2))  # Output: Same memory address as list1
```

### 3. What is the output of the following code and why ? 
a = 10
b = 10
print(a is b)

##### 
- *is* Operator: Checks whether two variables reference the same object in memory.
- Interning: Python reuses the same object for small integers (typically between -5 and 256) and some strings to save space and improve performance.

```python
a = 10
b = 10
print(a is b)  # Output: True
```

#### 
- When we assign a = 10, Python creates an integer object with the value 10 and assigns it to the variable a.
- When we assign b = 10, Python reuses the existing integer object with the value 10 and assigns it to the variable b.
Since both a and b reference the same object in memory, a is b returns True.

### 4. Print(0.1+0.2 == 0.3)

```python
print(0.1+0.2 == 0.3) # False
print(0.1+0.2) # 0.30000000000000004
```

#### Floating-Point Representation:

- Computers represent floating-point numbers using a fixed number of binary digits (bits). This representation can lead to small rounding errors because some decimal numbers cannot be represented exactly in binary form.
- For example, the decimal number 0.1 cannot be represented exactly as a binary fraction, so it is approximated.

#### 2. Precision Errors:

- When we perform arithmetic operations with floating-point numbers, these small errors can accumulate, leading to results that are slightly off from the expected value.
- In our example, 0.1 + 0.2 results in 0.30000000000000004 instead of 0.3 due to these precision errors.

```python
print(0.1*0.2 == 0.02) # False
print(0.1*0.2) # 0.020000000000000004
```

#### Pass by Value
1. When a function is called, a copy of the actual value is passed to the function.
2. Changes made to the parameter inside the function do not affect the original variable.
3. Immutable data types in Python (e.g., integers, floats, strings, tuples) behave like they are passed by value because their values cannot be changed.

#### Pass by Reference
1. When a function is called, a reference to the actual memory address of the variable is passed.
2. Changes made to the parameter inside the function affect the original variable.
3. Mutable data types in Python (e.g., lists, dictionaries, sets) behave like they are passed by reference because their contents can be modified.

```python
# Pass by value
def modify_value(x):
    x = x + 1
    print("Inside function:", x)

a = 10
modify_value(a)
print("Outside function:", a)
```

```python
# Pass by Reference
def modify_list(lst):
    lst.append(4)
    print("Inside function:", lst)

my_list = [1, 2, 3]
modify_list(my_list)
print("Outside function:", my_list)
```
