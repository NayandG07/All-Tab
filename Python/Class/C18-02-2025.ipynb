## Qs. What happens when we apply split() in "Good Morning Everyone" ?

- ##### When we apply the split() method to the string "Good Morning Everyone", the method looks for whitespaces in the string. Then the method splits the string at each whitespaces resulting in three separate substrings; i.e. "Good", "Morning" & "Everyone", which are then collected into a list ["Good", "Morning", "Everyone"].

```python
s = "Good Morning Everyone"
s.split()
```
