## Dictionary -
- ### Dictionaries are used to store Data Values in Key-Value pairs. 
- ### Dictionaries are collections of the pairs which is ordered, mutable and do not allow duplicates of keys. (values can be duplicated)

```python
a = {10:'Rajiv', 20:'Vipul', 'Raja':3000}
# a = {10:'Rajiv', 20:'Vipul', 10:'Raja'}
print(type(a), len(a))

print(a.get(20))
print(a[20][1]) # a[key][index] # Referenced indicing
```

```python
a1 = {10:'Rahul', (10,20,30):'Vipul', 200:'Raja'}
a1[40] = 'Utkarsh'
print(a1)

print(a1.keys()) # 
print(a1.values())
print(a1.items()) # Returns a list of (key,value)tuples.
```

```python
x = [10,20,'Vipul']
y = ['Rahul', 'Harry', 300]

print(dict(zip(x,y))) # zip() function

print(tuple(zip(x,y)))

d = {}
d[x[0]] = y[0]
d[x[1]] = y[1]
d[x[2]] = y[2]
print(d)
```

```python
a = {'a':10, 'b':20, 'c':30}
print(a.pop('b'))
a.update({'d':40, 'e':50})
print(a)
```

```python
car = {"Brand":"Ford", "Model":"Mustang", "Year":1964}
print(car.get("Brand"))
print(car.items())
car.update({'Color':'White'}) # To add a new key-value pair
print(car)

car.clear() # Empty the dictionary
print(car) # Prints : {}
```

### insert() 7000 after 6000.

```python
# Define the nested tuple
a = (10, 20, 30, (300, 400, (5000, 6000), 500), 400)

# Converted the nested tuple to a list, modified it, and converted it back to a tuple
a_list = list(a)
a_list[3] = list(a_list[3])
a_list[3][2] = list(a_list[3][2])

a_list[3][2].insert(2, 7000) # Insert 6000 at index 1 in the innermost list

a_list[3][2] = tuple(a_list[3][2])
a_list[3] = tuple(a_list[3])
A = tuple(a_list)

print(A)
```

### Qs.3. WAP to accept marks of 6 students and display them in a sorted manner. 

```python
marks = []

for i in range(1,6): # initialized with 1 to 6
    l2 = float(input("Enter marks for student " + str(i)+": "))
    marks.append(l2)

marks.sort()

print("Marks are: ", marks)
```
