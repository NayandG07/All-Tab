### WAP to find the keys with the maximum and minimum values in dictionary.

```python
d = {"vipul" : 89, "Siya" : 70, "Sam" : 62, "Rosy" : 22, "Payal" : 100}
a = d.values()
max_val = max(a)
min_val = min(a)
k = d.keys()
for i in k:
    # if d[i] == max_val:
    #     print(i, max_val)
    # if d[i] == min_val:
    #     print(i, min_val)
    if d[i] == max_val or d[i] == min_val:
        print(i, d[i])
```

```python
d = {"vipul" : 89, "Siya" : 70, "Sam" : 62, "Rosy" : 22, "Payal" : 100}
max_val = max(d, key=lambda x: d[x])
min_val = min(d, key=lambda x: d[x])
print("Key with max value: ",max_val,d[max_val])
print("Key with min value: ",min_val,d[min_val])
```

## OOPs

#### OOPs is a programming paradigm that offers a way to organise programs such that the traits and actions are encapsulated within individual objects. For instance, an object may stand for a person with attributes like name, age and addresses and actions like walking, running breathing etc. Alternatively, it may stand for an email with attributes such as recepient list, subject and body, and actions like adding attachments and sending the emails. 
#### In the other way, OOP is a way methodology for encoding concrete, physical things such as cars and relationship between companies and staff. Real world things are represented in OOP as software objects that contain some data and are able to execute some operations. Some points abou classes which is used in Python -
- Classes are created by the Key word 'Class'.
- Attributes are the varibales that belong to a class.
- Attributes are always public and can be accessed using the dot operator.

```python
class One:
    x = 20
    y = 10

obj = One()
print(obj.x)
obj1 = One()
print(obj1.y)
```
