#### What is a File?
- A File is a named location on disk to store related information permanantely. It is used to store data that can be retreived and used even after teh program is terminated. 

#### What is File handling?
- Python File Handling involves creating, openning, reading, writing, and closing the file to store data permanently on a hard disk beyond a programs temporary RAM storage. It enabes easy management of user details, logs, and large data sets, making it essential for persistent data storage and effiecient program execution through built in functions and methods.

#### **Need for File Handling** - 
- 1. ***Permanent Storage*** - Data which is stored in RAM is voaltile. Files provide permanent storage on a disk such that data remains after the termination of the program. Ex- Storing student results or User preferences.
- 2. ***Data Sharing*** - Files facilitate programs to exchange data between systems or between different progams. Ex- A Payroll system storing a report that can be read and utilised by another program.
- 3. ***Management of Large Data*** - Files can hold large amounts of Data that can not in fit in memory, making it possible to efficiently manage millions of records or large files. Ex- A banks transaction history.
- 4. ***Continuity of Programs*** - Files make it possible for a program to pick up where it left off without having to start all over again. Ex- Progress of a video game saving to allow users to resume from where they left.
- 5. ***Everyday Applications*** - File store user information logs, configuration and large data. Ex- User login deatils, system login or saving configuratuions.
- 6. ***Practical Needs*** - Without handling files the users would be required to enter their data whenever they run the program. Thus, making the program impractical and inefficient. Ex- Having to re-enter information each time logging onto a site is irritating.

- In short, File handling is important to enable the construction of useful, efficient and long term program in which information must be saved and reused.

#### ***readline()***
- ***Purpose***: Reads a single line from the file.
- ***Behavior***: Each call to readline() reads the next line in the file until the end is reached.
- ***Return Type***: Returns a string containing the line, including the newline character (\n) at the end (if present).

#### ***readlines()***
- ***Purpose***: Reads all the lines from the file and returns them as a list of strings.
- ***Behavior***: Each element in the list corresponds to a line in the file, including the newline character (\n) at the end of each line (if present).
- ***Return Type***: Returns a list of strings.

#### ***tell() method*** - 
- ***Return Value***: It returns an integer representing the current position of the file pointer in bytes from the beginning of the file.
- ***Usage***: Useful for tracking where you are in the file, especially when working with large files or when you need to resume reading/writing from a specific position.

```python
def count_lines(filename):
    with open(filename, 'r') as file:
        count = 0
        for line in file:
            count += 1
    return count

filename = 'demo.txt'
line_count = count_lines(filename)
print(f"Number of lines in {filename}: {line_count}")
```

#### Qs. WA function to count uppercase characters in a file.

```python
def count_uppercase_characters(file_name):
    file = open(file_name, "r")
    content = file.read()
    file.close()

    uppercase_count = sum(1 for i in content if i.isupper())
    return uppercase_count

file_name = "demo.txt"
uppercase_count = count_uppercase_characters(file_name)
print(f"Number of uppercase characters in the file: {uppercase_count}")
```

```python
def count_words(filename):
    count = 0
    with open(filename, 'r') as file:
        for i in file:
            words = i.split()
            count += len(words)
    return count

file_path = "demo.txt"
word_count = count_words(file_path)
print(f"Total number of words in '{file_path}': {word_count}")
```
