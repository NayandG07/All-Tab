### Qs. WAP that takes 3 side lengths as input and determines whether they form a valid triangle or not. If valid, classify the triangle as Equilateral, Isosceles or Scalene, using a nested if statement. 

```python
a = int(input("Enter side 1: "))
b = int(input("Enter side 2: "))
c = int(input("Enter side 3: "))

if a + b > c and a + c > b and b + c > a:
    print("The triangle is valid.")

    if a == b == c:
        print("The triangle is equilateral.") # 7 7 7
    elif a == b or a == c or b == c:
        print("The triangle is isosceles.") # 7 7 8
    else:
        print("The triangle is scalene.")
else:
    print("The triangle is not valid.")
```

### Password

```python
password = input("Enter password: ")
is_upper = is_lower = is_digit = is_special = False
special_chars = "@#$%&"
if len(password) >= 8:
    for i in password:
        if i.isupper():
            is_upper = True
        elif i.islower():
            is_lower = True
        elif i.isdigit():
            is_digit = True
        elif i in special_chars:
            is_special = True
if (is_upper and is_lower and is_digit and is_special):
    print("Strong password!")
else:
    print("Weak password.")
```

### Hollow Diamond

```python
n = int(input("Enter the number of rows: "))

# Upper part of the hollow diamond
for i in range(n):
    if i == 0:
        print(" " * (n - i - 1) + "*")
    else:
        print(" " * (n - i - 1) + "*" + " " * (2 * i - 1) + "*")

# Lower part of the hollow diamond
for i in range(n - 2, -1, -1):
    if i == 0:
        print(" " * (n - i - 1) + "*")
    else:
        print(" " * (n - i - 1) + "*" + " " * (2 * i - 1) + "*")
```

### Armstrong Numeber

```python
num = int(input("Enter a number: "))
sum = 0
temp = num
num_digits = len(str(num))

while temp > 0:
    digit = temp % 10
    sum += digit ** num_digits
    temp //= 10

if num == sum:
    print(f"{num} is an Armstrong number")
else:
    print(f"{num} is not an Armstrong number")
```

### Fibonacci Series

```python
def fibonacci_recursive(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

n = int(input("Enter the number of terms: "))
for i in range(n):
    print(fibonacci_recursive(i), end=" ")
print()
```

### Palindrome

```python
num = int(input("Enter a number: "))
rev = 0
temp = num
while num != 0:
    digit = num % 10
    rev = rev*10 + digit
    num //= 10

if temp != rev:
    print(f"{temp} is not a palindrome")
else:
    print(f"{temp} is a palindrome")
```

```python
num = int(input("Enter a number: "))
original_num = num
reverse_num = 0

while num > 0:
    digit = num % 10
    reverse_num = reverse_num * 10 + digit
    num //= 10
else:
    # This block executes after the while loop completes
    if original_num == reverse_num:
        print(f"{original_num} is a palindrome.")
    else:
        print(f"{original_num} is not a palindrome.")
```

### Collatz Conjecture
- The Collatz Conjecture is a mathematical conjecture that states that for any positive integer n, the following process will eventually reach the number 1:

- If n is even, divide it by 2 (i.e., n = n / 2).
- If n is odd, multiply it by 3 and add 1 (i.e., n = 3 * n + 1).
- Repeat the process with the new value of n.
- The conjecture claims that no matter what positive integer you start with, you will always eventually reach 1.

```python
def collatz_conjecture(n):
    while n != 1:
        print(n, end=" -> ")
        if n % 2 == 0:  # If n is even
            n = n // 2
        else:  # If n is odd
            n = 3 * n + 1
    print(1)  # Print the final 1

# Input from the user
num = int(input("Enter a positive integer: "))
if num > 0:
    collatz_conjecture(num)
else:
    print("Please enter a positive integer.")
```

### Pascal's Triangle

```python
def pascals_triangle(n):
    for i in range(n):
        # Print spaces for alignment
        print(" " * (n - i - 1), end="")
        
        # Calculate and print each value in the row
        value = 1
        for j in range(i + 1):
            print(value, end=" ")
            value = value * (i - j) // (j + 1)
        
        # Move to the next line
        print()

# Input from the user
rows = int(input("Enter the number of rows for Pascal's Triangle: "))
pascals_triangle(rows)
```

### Armstrong number between 1 to 1000, skips those which are not, stops after 10 numbers.

### Prime from 1 to 1000, skips even numbers and uses square root for efficiency.

```python
import math

def find_primes(limit):
    primes = []
    for i in range(2, limit + 1):
        if i > 2 and i % 2 == 0:  # Skip even numbers greater than 2
            continue
        
        is_prime = True
        for j in range(2, int(math.sqrt(i)) + 1):  # Check divisors up to the square root of the number
            if i % j == 0:
                is_prime = False
                break
        
        if is_prime:
            primes.append(i)
    
    return primes

# Find all prime numbers between 1 and 1000
primes = find_primes(1000)
print("Prime numbers between 1 and 1000:")
print(primes)
```
