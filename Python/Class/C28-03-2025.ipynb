```python
def mod_immutable(x):
    x = 100  # Reassigning x does not affect the original variable
    print("Inside function (Immutable):", x)

num = 10
mod_immutable(num)
print("Outside function (Immutable):", num)  # Original value remains unchanged

def mod_mutable(x):
    x.append(4)  # Modifies the original list
    print("Inside function (Mutable):", x)

my_list = [1, 2, 3]
mod_mutable(my_list)
print("Outside function (Mutable):", my_list)  # Original value is modified
```

### Write a function that returns the absolute value of a number.

```python
def absolute_value(num):
    if num < 0:
        return -num
    return num

# Test cases
print(absolute_value(-5))  # Output: 5
print(absolute_value(10))  # Output: 10
print(absolute_value(0))   # Output: 0
```

### Write a function is_Prime() that returns a 1 if the argument is passed to it is a prime number, and a 0 otherwise.

```python
def is_Prime(num):
    if num <= 1:
        return 0
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return 0
    return 1

# Test cases
print(is_Prime(2))   # Output: 1 (Prime)
print(is_Prime(4))   # Output: 0 (Not Prime)
print(is_Prime(17))  # Output: 1 (Prime)
print(is_Prime(1))   # Output: 0 (Not Prime)
print(is_Prime(0))   # Output: 0 (Not Prime)
```
