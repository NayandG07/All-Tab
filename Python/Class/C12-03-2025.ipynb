### Qs. WAP to take n integers as input from the user, then store them in a list and print the final list. 

```python
num = []
n = int(input("Enter the number of elements: "))

for i in range(n):
    num.append(int(input(f"Enter element {i}: ")))
print("Numbers: ",num)
```

### Qs. Perform addition in a list a = [10,20,30,40,50] using for loop.

```python
a = [10, 20, 30, 40, 50]
sum = 0
for i in a:
    sum += i
print("Sum of the elements of the list:",sum)
```

```python
n = int(input("Enter a number: "))
factorial = 1
for i in range(1,n+1):
    factorial *= i
print("Factorial of",n, " is: ",factorial)
```
