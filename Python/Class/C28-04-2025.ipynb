#### Calculate the percentage of Physics, Chemistry & Maths using Student class.

### @property decorator - The @property decorator in Python is used to define a method as a property, allowing you to access it like an attribute without explicitly calling it as a method. This is useful for encapsulating logic while maintaining a clean and intuitive interface.

```python
class Student:
    def __init__(self, physics, chemistry, math):
        self.physics = physics
        self.chemistry = chemistry
        self.math = math
        # self.total = (physics + chemistry + math) / 3

    @property
    def total(self):
        percentage = self.physics + self.chemistry + self.math
        return percentage / 3
    
s1 = Student(90, 80, 85)

print("Percentage of student 1:", s1.total)

s1.physics = 95
print(s1.total)
```

#### @classmethod - The *@classmethod* decorator in Python is used to define a method that is bound to the class and not the instance of the class. It can be called on the class itself or on an instance, and it takes the class (cls) as its first argument.

```python
class Car:
    base_amt = 1000000
    def __init__(self, window, doors):
        self.window = window
        self.doors = doors
    
    def base_price(self):
        return f"Base price is: {self.base_amt}"
    
car1 = Car(5,4)
car1.base_amt # By creating an instance
Car.base_amt # Directly accessing the class variable using the class name
Car.base_price(car1) # Passing the instance of the class as an argument to the method
```

```python
class Circle:
    pi = 3.14
    @classmethod
    def area(cls, radius):
        return cls.pi * radius * radius
    
Circle.area(5) # Calling the class method using the class name
area = Circle.area(5) # Calling the class method using the instance of the class
print("Area of the circle is: ", area)
```

```python
class Cylinder:
    pi = 3.14
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
    
    @classmethod
    def volume(cls, radius, height):
        return cls.pi * radius * radius * height
    
Cylinder.volume(5, 10) # Calling the class method using the class name
volume = Cylinder.volume(5, 10) # Calling the class method using the instance of the class
print("Volume of the cylinder is: ", volume)
```

### Operator and Dunder Functions: 

```python
class Complex:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag
    
    def showNum(self):
        print(f"{self.real} + {self.imag}j")

    def add(self, other):
        return Complex(self.real + other.real, self.imag + other.imag)
    
    def __add__(self, n2):
        R = self.real + n2.real
        I = self.imag + n2.imag
        return Complex(R, I)
    

n1 = Complex(2, 3)
n1.showNum()

n2 = Complex(4, 5)
n2.showNum()

n3 = n1 + n2 # n1.__add__(n2)
n3.showNum()

n4 = n1.add(n2)
n4.showNum()
```

#### Define a Circle class to create with radius r using the constructor. Define an Area() method of the class which calculates the area of the circle. Define a perimeter() method of the class whicő allows you to calculate the perimeter of the circle.

```python
class Circle:
    pi = 3.14
    def __init__(self, radius):
        self.radius = radius
    
    @classmethod
    def area(cls, radius):
        return cls.pi * radius ** 2
    
    @classmethod
    def perimeter(cls, radius):
        return 2 * cls.pi * radius
    
circle1 = Circle(5)
print("Area of the circle is: ", Circle.area(circle1.radius))
print("Perimeter of the circle is: ", Circle.perimeter(circle1.radius))
```

#### Define an Employee class with attributes Role, Dept, and salary and show the details using a showDetails method. Create and Engineer class that Inherites properies from Employee class that has Additional attributes nam and age.

```python
class Employee:
    def __init__(self, role, dept, salary):
        self.role = role
        self.dept = dept
        self.salary = salary

    def showDetails(self):
        return f"Role: {self.role}, Department: {self.dept}, Salary: {self.salary}"

class Engineer(Employee):
    def __init__(self, role, dept, salary, name, age):
        super().__init__(role, dept, salary)
        self.name = name
        self.age = age

    def showDetails(self):
        details = super().showDetails()
        return f"{details}, Name: {self.name}, Age: {self.age}"

emp = Employee("Manager", "HR", 50000)
print(emp.showDetails())

eng = Engineer("Software Engineer", "IT", 70000, "Alice", 30)
print(eng.showDetails())
```
