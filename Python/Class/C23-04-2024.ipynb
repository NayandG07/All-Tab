#### **Encapsulation** --> Encapsulation is the process of bundling data (attributes) and methods (functions) that operate on the data into a single unit, i.e., a class. It also involves restricting direct access to some of the object's components to enforce data hiding.
#### **Abstraction** --> Abstraction is the concept of hiding the implementation details of a class and exposing only the essential features. It focuses on what an object does rather than how it does it. In Python, abstraction is often achieved using abstract base classes (ABC).

```python
class Car:
    def __init__(self):
        self.acc = False
        self.brk = False
        self.clt = False

    def start(self):
        self.clt = True
        self.acc = True
        print("Car gets started.")

c1 = Car()
c1.start()
```

#### **del keyword** --> Deletes a variable from the local or global namespace. Delete the properties od the whole object itself.

```python
class Student():
    def __init__(self, name):
        self.name = name

s1 = Student("John")
print(s1.name)

del s1.name
print(s1.name)  # This will raise an AttributeError since name has been deleted
```

#### **Access Specifiers**:
#### - 1. Public: Members declared as public can be accessed from any other class or module. Declared without any prefix.
#### - 2. Private: Members declared as private can only be accessed within the same class. Declared with the prefix "private".
#### - 3. Protected: Members declared as protected can be accessed within the same class and by derived class instances. Declared with the prefix "protected". 

```python
class Student():
    def __init__(self, name):
        self.name = name
        self.__roll  = 47
    
    def get_roll(self):
        return self.__roll

    def set_roll(self, roll):
        self.__roll = roll
    
s1 = Student("John")
print(s1.get_roll())  # Accessing private variable using getter method
```

#### Qs. 1 Create a Library book class with attributes book_id, title, author and status (available/borrowed). Implement method to: 
#### - 1. Borrow a book (if avaialable)
#### - 2. Return a book (if borrow)
#### - 3. Display book details.

```python
class LibraryBook:
    def __init__(self, book_id, title, author, status="available"):
        self.book_id = book_id
        self.title = title
        self.author = author
        self.status = status

    def borrow_book(self):
        if self.status == "available":
            self.status = "borrowed"
            print(f"The book '{self.title}' has been borrowed.")
        else:
            print(f"The book '{self.title}' is already borrowed.")

    def return_book(self):
        if self.status == "borrowed":
            self.status = "available"
            print(f"The book '{self.title}' has been returned.")
        else:
            print(f"The book '{self.title}' is not currently borrowed.")

    def display_details(self):
        print(f"Book ID: {self.book_id}")
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print(f"Status: {self.status}")


book = LibraryBook(1, "Python Programming", "John Doe")

book.display_details()

book.borrow_book()

book.borrow_book()

book.return_book()

book.return_book()

book.display_details()
```
