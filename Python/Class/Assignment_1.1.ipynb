## Beginner Level Questions



### 1. isalpha() and isalnum()

Check if the string "Hello123" is purely alphabetic or alphanumeric.

```python
s = "Hello123"
print(s.isalpha())
print(s.isalnum())
```

### 2. index() and find()

Find the position of the substring "world" in the string "Hello world, welcome to the world of Python" using both methods.

```python
s = "Hello world, welcome to the world of Python"
print(s.find("Python"))
print(s.index("Python"))
```

### 3. capitalize() and islower()

Capitalize the first letter of the string "python programming" only if it's entirely in lowercase.

```python
s1 = "python programming"
if s1.islower():
    print(s1.capitalize())
```

### 4. isdigit() and isnumeric()

Verify if the string "²³" is considered numeric by Python.

```python
s2 = "23"
print(s2.isdigit())
print(s2.isnumeric())
```

### 5. endswith() and startswith()

Check if the string "example.py" starts with "ex" and ends with ".py".

```python
s = "example.py"
print(s.startswith("ex"))
print(s.endswith(".py"))
```

### 6. split() and join()

Split the sentence "apple,banana,cherry" by commas and then join the list back into a single string separated by semicolons.

```python
s = "apples, banana, cherry"
print(s.split(", "))
print("; ".join(s.split(", ")))
```

### 7. replace() and swapcase()

Replace all spaces in "Hello World" with underscores and then swap the case of all characters.

```python
s = "Hello World!"
s1 = s.replace(" ", "_")
s1.swapcase()

# s.replace(" ", "_").swapcase()
```

### 8. append() and extend()

Append the list [4, 5] to [1, 2, 3] using append() and then do the same using extend(). Observe the difference.

```python
a = [1, 2, 3]
b = [4, 5]
a.append(b)
print(a)
```

```python
a = [1, 2, 3]
b = [4, 5]
a.extend(b)
print(a)
```

### 9. sort() and reverse()

Sort the list [4, 2, 3, 1] in ascending order and then reverse the sorted list.

```python
l1 = [4, 2, 3, 1]
l1.sort()
print(l1)

l1.reverse()
print(l1)
```

### 10. insert() and pop()

Insert the element 100 at the second position of the list [10, 20, 30] and then remove the last element using pop().

```python

l2 = [10, 20, 30]
l2.insert(1, 100)
l2.pop()
l2
```
