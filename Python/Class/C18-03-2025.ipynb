```python
marks = [10, 30, 40, 50]
new_marks = []

for i in marks:
    new_marks.append(i + 2)

print(new_marks)
```

```python
marks = [10, 30, 40, 50]
for i in range(len(marks)):
    marks[i] += 2

print(marks)
```

```python
marks = [10, 30, 40, 50]
new_marks = [i + 2 for i in marks]

print(marks)
```

```python
# Cubes of even numbers
Cubes = []
for i in range(11):
    if i % 2 == 0:
        Cubes.append(i ** 3)

print(Cubes)
```

```python
Cubes = [i ** 3 for i in range(11) if i % 2 == 0]
print(Cubes)
```

```python
# Words which has 'a' in it
l1 = ["ant", "bat", "crisp", "man"]
new_list = []

for i in l1:
    if "a" in i:
        new_list.append(i)

print(new_list)
```

```python
l1 = ["ant", "bat", "crisp", "man"]
new_list = [i for i in l1 if "a" in i]
print(new_list)
```

### Qs. temps = [10, 33, 19, 15, 25]
### if temp <= 15 --> cold
### else: ---> warm
### temp = [cold, warm, warm, cold, warm]

```python
temps = [10, 33, 19, 15, 25]

temp = ["cold" if i <= 15 else "warm" for i in temps]
print(temp)
```

```python
temps = [10, 33, 19, 15, 25]
temp = []
for i in temp:
    if i > 20:
        temp.append("warm")
    else:
        temp.append("cold")

print(temp)
```
