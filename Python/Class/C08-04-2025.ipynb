#### Map Function- The map() function in Python is used to apply a given function to all items in an iterable (like a list or tuple) and return a map object (an iterator). next() function to iterate tht next element.

```python
num = [1,2,3,4,5,7]
def multiple(items):
    return items * 5

result = map(multiple, num)
# print(list(result)) # Output: [5, 10, 15, 20, 25, 35] 
print(next(result)) # print the first element of the iterator

print(next(result)) # print the second element of the iterator

print(list(result))
```

```python
def multiple(items):
    return items ** 2

result = map(multiple, num)
print(list(result))
```

```python
def myfunc(n):
    return len(n)

x = map(myfunc, ('apple', 'banana', 'cherry'))
print(list(x))
```

#### The filter() function in Python is used to filter elements from an iterable based on a condition defined in a function. It returns an iterator containing only the elements for which the function returns True

```python
num = [10,20,21,32,40,15]
def a(i):
    return i<25

result = filter(a, num)
print(list(result))
```

```python
marks = [32,67,43,50,55,78]
def check(m):
    return m >= 40

result = filter(check, marks)
print(list(result))
```

```python
numbers = [1,2,3,4,5]
def is_odd(n):
    return n % 2 != 0

odd_num = filter(is_odd, numbers)
print(next(odd_num))
print(list(odd_num))
```

### Lambda function: The lambda function in Python is an anonymous (nameless) function defined using the lambda keyword. It is often used for short, simple operations where defining a full function is unnecessary.
- Always use map(),filter(),shorted()

#### lambda arg(multi): expressions

```python
num = [10,20,30,40]
# result = map(lambda x: x > 20, num) # Output: [False, False, True, True]
result = filter(lambda x: x > 20, num) # Output: [30, 40]
print(list(result))
print(result)
```

```python
def add(a,b,c):
    return a + b + c

print(add(4,5,6))

result = map(lambda x,y,z: x + y + z, [4], [5], [6])

print(list(result))
```

```python
sum = lambda a,b,c: print(a+b+c)
sum(4,5,6)
```

### WAP to sort a list of state names based on the length of each name. 

```python
states = ["Sikkim", "Manipur", "Diu", "Chandigarh", "Delhi"]
sorted_states = sorted(states, key=lambda x: len(x))
print(sorted_states)
```

```python
states = ["Sikkim", "Manipur", "Diu", "Chandigarh", "Delhi"]
def length(s):
    return len(s)

state = sorted(states, key=length)
print(state)
```

```python
states = ["Sikkim", "Manipur", "Diu", "Chandigarh", "Delhi"]
sort = sorted(states, key=lambda x: len(x), reverse=True)
print(sort)
```
