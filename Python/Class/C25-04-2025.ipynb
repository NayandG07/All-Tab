### Inheritance
- When one class(child/derived) derives the properities & methods of another class(parent/base)
- Syntax:::::
   class Parent:

   class Child(Parent):

```python
class A:
    @staticmethod
    def book():
        print("Mathematics")

    @staticmethod
    def note():
        print("Handwritten")

class E(A):
    def __init__(self, marks):
        self.marks = marks

e1 = E(460)
e2 = E(422)
#e1.marks
e2.book()
e2.note()
```

```python
#Single Inheritance
class Car:
    @staticmethod
    def start():
        return("Car started..")

    @staticmethod
    def stop():
        print("Car Stopped")

class Tata(Car):   # Inheriting the Parent Class
    def __init__(self, name):
        self.name = name

c1 = Tata("Curvv")
c2 = Tata("Altroz")
print(c1.name)
print(c1.start())

# you can also add more features as attributes like color, model.
```

```python
'''Create a Python program to model the following scenario using Hybrid Inheritance:

Vehicle class has a method vehicle_info().

Car class inherits from Vehicle and adds a method car_info().

Bike class also inherits from Vehicle and adds a method bike_info().

ElectricBike class inherits from both Bike and Car, and adds a method electric_info().

Tasks:

Create objects and call all available methods from ElectricBike.

'''
```

```python
# Base class
class Vehicle:
    def vehicle_info(self):
        print("This is a Vehicle.")

# First child class
class Car(Vehicle):
    def car_info(self):
        print("This is a Car.")

# Second child class
class Bike(Vehicle):
    def bike_info(self):
        print("This is a Bike.")

# Derived class using Hybrid Inheritance
class ElectricBike(Bike, Car):
    def electric_info(self):
        print("This is an Electric Bike.")

# Create an object of ElectricBike
ebike = ElectricBike()

# Calling methods
ebike.vehicle_info()    # Inherited from Vehicle
ebike.car_info()        # Inherited from Car
ebike.bike_info()       # Inherited from Bike
ebike.electric_info()   # Defined in ElectricBike
```

### Operator and Dunder function
- These methods provide a way to define specific behaviors for built-in operations or functionalities in Python classes.
- Also known as magic methods or special methods

```python

a + b # addition---> a.__add__(b)
a - b # subtraction---> a.__sub__(b)
a * b # multiplication---> a.__mul____(b)
a/b #division ---> a.__truediv____(b)
a%b # Mod ---> a.__mod____(b)
```

```python
class Complex:
    def __init__(self, real, img):
        self.real = real
        self.img = img

    def showNum(self):
        print(self.real,"i","+", self.img,"j")

    def add(self, n2):   # we manually created an add function
        newR = self.real + n2.real
        newI = self.img + n2.img
        return Complex(newR, newI)

n1 = Complex(2, 3)
n1.showNum()

n2 = Complex(4, 2)
n2.showNum()

n3 = n1.add(n2)
n3.showNum()

# Want directly to print
#print(n1+n2)
```

```python
class ComplexNumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def __add__(self, other):
        # Overloading the + operator
        return ComplexNumber(self.real + other.real, self.imag + other.imag)

    def __str__(self):
        # Overloading the str() function for better display
        return f"{self.real} + {self.imag}i"

# Creating two complex number objects
num1 = ComplexNumber(3, 5)
num2 = ComplexNumber(4, 7)

# Adding the two objects using the overloaded + operator
result = num1 + num2

# Displaying the result
print("The sum of the complex numbers is:", result)
```

```python
class Shape:
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width


circle = Circle(5)
rectangle = Rectangle(4, 6)

#Polymorphism
shapes = [circle, rectangle]
for shape in shapes:
    print(f"Area of {type(shape).__name__}: {shape.area()}")

```

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2  # or Set pi = (22/7)

    def perimeter(self):
        return 2 * 3.14 * self.radius

c1 = Circle(21)
print("The area of the circle is: ", c1.area())
print("The perimeter of the circle is: ", c1.perimeter())
```

```python
''' Create a class called Order which stores item & its price.
        Use Dunder function__gt__() to convey that:
            order1 > order2 if price of order1 > price of order2
'''
```

```python
class Order:
    def __init__(self, item, price):
        self.item = item
        self.price = price

    def __gt__(self, o2):   # dunder function for greater than function
        return self.price > o2.price

o1 = Order("Waffers", 40)
o2 = Order("Tea", 15)

print(o1 > o2)
```
